#### 1.11 Prove $n^{4}+4$ is composite for $n > 1$. 


An approach would be to factor this expression, for example: $n^{4}+4 = (n^2+an+2)(n^2+bn+2)$. 
Happily this works out using $a=2$ and $b=-2$. Both resulting factors are positive 
for all values of $n > 1.\;\;$ &#x2610;

#### 1.12 a) Prove or provide a counterexample: $a^n | b^n \implies a | b$


For $n = 1$ the assertion is true. Continuing with $n > 1$:
$a^n | b^n \implies a \cdot a^{n-1} \cdot c = b \cdot b^{n-1}$ with $c > 0$.
This means that either $a | b$ or $a | b^{n-1}$.
If $a | b$ the assertion is true. Taking $a \nmid b$
implies $a \ne 1$ and this
leads to a contradiction when we continue casting out factors of $b$ to reduce
the exponent to $0$: $a$ must divide in succession:
$b^{n-2},\; b^{n-3},\; \dots \; b^{1}, \; 1$.
This last step $a|1$ implies $a=1.\;\;$ &#x2610;


#### 1.12 b) Prove or provide a counterexample: $n^n | m^m \implies n | m$. 


This assertion is false. The premise implies $n \le m$. Suppose that $n \nmid m$ even though all
the prime factors of $n$ are also prime factors of $m$. For example $25$ and $55$. 
That statement concerns $n$ and $m$ without exponentiation. 


Now let's transition to $m^m$. This can stack up a lot of prime factors, eventually reaching or exceeding
the number of prime factors found in $n^n$. Whereupon it will be the case that $n^n | m^m$. 
Calculations follow for three examples. In the first case $4 \nmid 10$ but $10^{10}$ contains
$10$ factors of $2$ exceeding the $8$ found in $4^4$; so $4^4 \mid 10^{10}$ and 
this is a counterexample. $\otimes$

See the **`examples.ipynb`** notebook.

#### 1.12 c) For $n > 1$ show $a^n | (2b^n) \implies a|b$. 


If $a$ is odd we have $a^n|b^n$ and **1.12a** establishes the implication $a|b$.


If $a$ is even then suppose $a=2^s \cdot d$ and so $2^{ns} \cdot d^n | 2 \cdot b^n$. Dividing out one $2$ we have for some $m$:


$$b^n = 2^{ns-1}\cdot d^n \cdot m = 2^{ns-n+n-1}\cdot d^n \cdot m = 2^{n(s-1)+(n-1)}\cdot d^n \cdot m$$


Now the finesse: $b$ is raised to the $n$ so the right side must be *something* raised to the $n$. $d$ is odd so as 
2^{n(s-1)+(n-1)} is not some power of two raised to the $n$ (because $n-1 > 0$) we must conclude that $m$ is even.
Moving a factor of two from $m$ to the $2$ expression we have


$$b^n = 2^{ns} \cdot d^n \cdot m/2$$

Taking the $n$'th root of $m/2$ is not a problem; we could reverse it and require $m = 2 \cdot q^n$ for some $q$. 
So at this point we can take the $n$'th root of both sides to arrive at $b = a \cdot q$ and so $a | b.\;\;$ &#x2610;



#### 1.13 a) If $(a, b)=1$ and $(a/b)^m=n$ show that $b=1$.

In problem 1.7 we show that if the sum of two reduced fractions is an integer then their denominators are the same (or negatives). 
In this case we have $\frac{a^m}{b^m}-\frac{n}{1}=0$ and since $b$ is positive we have $b^m=1$ and therefore $b=1.\;\;$ &#x2610;

#### 1.13 b) If $n$ is not the $m$th power of a positive integer, show that $n^{1/m}$ is irrational.

Requiring $n > 1$ let us suppose the contrary, that $n^{1/m}=a/b$, a rational number.
Equivalently $n = \frac{a^m}{b^m}$. From 1.13a we have $b=1$ and $n=a^m$. Thusly
$n$ *is* the $m$th power of a positive integer, a contradiction.$\;\;$ &#x2610;

#### 1.14 If $(a, b) = 1$ and $ab=c^n$ show that $x$ and $y$ exist such that $a=x^n$ and $b=y^n$. [Hint: Consider $d=(a,c)$.]


Original remarks: 


- The hint: $d = (a,c) = au + cv$ and $e = (b,c) = bq + cr$. Hence $ab = c^n$.
- $c^n = \left( \frac{d-au}{v} \right)^{n} = ab$.


From Chip:

- Set up $a$ and $b$ in generalized prime product form using $p_i, \; e_i$ and $q_j, \; f_j$ for prime factors and their exponents.
- $ab = c^n$ is a big product of primes to powers.
- Take the $n$-th root of both sides; now $c$ is a product of primes to *fractional* powers with denominator $n$. 
- The resulting fractional exponents must be integers greater than or equal to $1$.
    - This takes a moment to think on: Violations would contradict the premise.
- This means $n$ divides every $e_i$ and every $f_j$; so we could say for example $f_j = n \cdot \delta_j$.
- Back to the product $a \cdot b$:
- $n | e_1$ means $e_1 = n \cdot \omega_1$ for some $\omega_1$ a positive integer.
- $p_1^{e_1}$ (the first component of $a$) can be written ${(p_1^{\omega_1})}^n$.
- The same is true for all $p_i$ and $q_j$ prime factors, distinct since $(a, b) = 1$.
- $p^\omega$ values are raised to the $n$ as are the $q^\delta$.
- In all cases (with everything raised to the $n$) one can roll up bases to form integers $x$ and $y$.
- Hence both $a$ and $b$ are $n$-th powers of two integers $x$ and $y.\; \; $ &#x2610;

#### 1.15 Show that every $n \ge 12$ is the sum of two composite numbers. 

$n$ even can be written $n = 4 + (n - 4) = 2\cdot 2 + 2 \cdot (\frac{n}{2}-2)$, a sum of two composites.

$n$ odd can be written $n = 9 + (n - 9)$ and noting $n - 9$ is even we have $n = 3\cdot 3 + 2\cdot(\frac{n-9}{2})$, again a sum of two composites.

This works for $n=13$ and $n=14$ and up.$\;\;$ &#x2610;

#### 1.16 Show that if $2^n-1$ is prime, $n$ is prime.

The appearance  of $x^\alpha-1$ is 
suggestive of a telescoping factorization.
Factors in turn imply composite; so this proceeds by way of 
contradiction: Assume there is some $n$ for which $2^n-1$ is
prime and $n$ is *not* prime, $n = ab$ for $a, b > 1$. 
Factor the prime number $2^{ab}-1$ and establish both factors 
are greater than one, a contradiction hence $n$ must be prime. 


Here is a factorization that falls from the sky (say):


$2^{ab}-1 = (2^a-1)(2^X + 2^{X-1} + 2^{X-2} + \dots + 2^{1} + 2^{0})$. 


The burden of proof here is to determine $X$ and show that the 
factorization with this $X$ does indeed give $2^{ab}-1$.


Set $X$ so the product of the two first terms $2^a \cdot 2^X$ is $2^{ab}$.


$\begin{align} X = ab - a = a(b-1) \end{align}$. 


Now both $2^a$ and $-1$ multiply distributively across the terms of the 
second factor. The products from $2^a$ cancel those from $(-1)$ but for 
the last, $-1 \cdot 1 = -1$, telescoping down to $2^{ab}-1$. 


As $a > 1$ we see the first factor is greater than $2$. 


Likewise since $a(b-1) > 1$ we see the second factor is a sum of positive integers
that will exceed $1$. Both the factors in this factorization are greater than one.
Therefore $2^n-1$ is not a prime when $n$ is not prime; so $n$ must be a prime.
Cue the box: $\;\;$ &#x2610;



However this is a bit of swindle as we have not shown that $n$ prime produces
or even guarantees that $2^n-1$ is prime. Clearly the proposition works for
$n = 2, \; 3, \; 5, \; 7$. That is, $3, \; 7, \; 31, \; \textrm{ and } 127$
are all prime... but this trend breaks down for $n = 11$. 
(These are Mersenne primes; and they show up again in problems 2.18 and 2.19.)

#### 1.17 Show that if $2^n + 1$ is prime, $n = 2^k$ for some $k \ge 0$.

This follows a similar approach to that used in **1.16** above; with the
telescoping factorization relying on alternating signs in the second factor. 
Here we suppose $n=2^k\cdot b$ with $b$ an odd number. The contradiction 
is reached by assuming $b \ge 3$ and factoring $2^n+1$ so as to imply it is
composite. The conclusion is then that $b = 1$ and $n$ is simply $2^k$.


Begin by presuming a factorization:


$2^n+1 = 2^{2^{k}b}+1=(2^{2^{k}}+1)\cdot(Y^X - Y^{X-1} + Y^{X-2} - \dots - Y^{1} + Y^{0})$


Now to sort both $Y$ and $X$; noting that the second factor must have
an odd number of terms for the alternating signs to resolve the telescope properly.
Noting the 'counting down' exponents $X, X-1, \dots, 1$: $X$ must
be even.


First term product of these factors is $2^n$ so 
$2^n = 2^{2^{k}b} = 2^{2^{k}}\cdot Y^X$ so $Y=2^{2^{k}}$ and $X=(b-1) \ge 2$.
$b$ is odd so $X$ is even as required in the factorization. Both
factors are greater than one; where the right factor might
require some staring. One approach is to pair the terms of the sum
as $positive + negative$ with a decreasing exponent. 
Example $Y^8-Y^7$ where $Y$ is a positive integer. The result is
that each pair has a net positive sign; plus the trailing $1$
giving a total $> 1$. 


As a result the factorization arrives at the
contradiction that $2^n+1$ is composite.$\;\;$ &#x2610;

#### 1.18 If $m > n$ compute the gcd $g = (a^{2^{m}} + 1, a^{2^{n}} + 1)$ in terms of $a$. 


***Hint: Define $A_n = a^{2^{n}}+1$ and show that $A_n | A_m - 2$.***


With this hint the problem is 'Give $(A_m, A_n)$ in terms of $a$'. If as the hint hints the number
$A_m - 2$ is divisible by $A_n$ then the only possible gcd values for $A_m$ and $A_n$ are $1$ and $2$.


As $m > n \ge 1$, consequently $2^m$ can be divided by $2^n$ an even number of times, in fact $2^{m-n}$ times. 
I resort to a telescoping product that follows machinery in problem 1.16.


First to follow the hint: Define $A_n$ as above and show $A_m - 2 = A_n \cdot \gamma$ (for some $\gamma$). 
Anticipating a telescoping product we suppose $\gamma$ will be a sequence of alternating-sign terms.


$A_m-2 = a^{2^m} +  1 - 2 = a^{2^m} -  1 = A_n \cdot \gamma = (a^{2^n} + 1) \cdot \gamma = (a^{2^n} + 1) \cdot (A - B + C \dots + Y - Z)$.


From telescope logic for the binomial $a^{2^n} + 1$ the first term $a^{2^n}$ multiplied
by some term $K$ in $\gamma$ will be the negative of the second term $1$ multiplied by term 
$J$ of $\gamma$ so that these sum to $0$.


It must be that $a^{2^n} \cdot A = a^{2^m}$, the non-telescoping lead term. The eyepiece?
So apparently $A$, the first term in $\gamma$, must be 
$\frac{a^{2^m}}{a^{2^n}} = a^{2^m} \cdot a^{-2^n} = a^{2^m-2^n}$.


Likewise $a^{2^n} \cdot B = A \cdot 1$, the first telescoping pair.


Also $a^{2^n} \cdot C = B$


...and so on until...


$a^{2^n} \cdot Z = Y$


and finally $Z = 1$, the second non-telescoping term (the lens!), which will be the $-1$ of $A_m - 2$.


This result $Z = 1$ gives $a^{2^n} = Y$. Hence the multiplying factor to go from one term
to the next in $\gamma$ must be $\frac{1}{a^{2^n}}$. I'll write this as $a^{-2^n}$.


$B=\frac{A}{a^{2^n}}=a^{2^m-2 \cdot 2^n} \textrm{ and } C=\frac{B}{a^{2^n}}=a^{2^m-3 \cdot 2^n}$.


That is, each successive term of $\gamma$ is the previous term divided by $a^{2^n}$; and this 
produces an even number of alternating-sign terms. The leading term being positive this means the final term
is $-1$, negative as required. The factorization works:


$A_m - 2 =  a^{2^m} - 1 = (a^{2^n} + 1)(a^{2^m-2^n} - a^{2^m-2 \cdot 2^n} + a^{2^m-3 \cdot 2^n} - a^{2^m-4 \cdot 2^n} + \dots + a^{-2^n} - 1)= (A_n + 1) \cdot \gamma$


We have gcd $g=(A_n, A_m) \textrm{ so } g|A_m$. While $A_n \nmid A_m$ we do have 
$A_n | A_m - 2$ and therefore $g | A_m - 2$ as well. So $g|A_m \textrm{ and } g|A_m-2$.
Hence $g$ can only be $1$ or $2$. If $a$ is odd then $A_n$ and $A_m$ will be even
and $g=2$. If $a$ is even then $A_n$ and $A_m$ will be odd and $g=1.\;\;$ &#x2610;

#### 1.19 Prove for the Fibonacci sequence that $(f_{n}, f_{n+1}) = 1$.

Note Theorem 1.1c) implies that $(a, b) = (a, b + na)$; so $(a, a + b) = (a, b)$. 
For the first several Fibonacci numbers 1, 1, 2, 3, 5, 8 etcetera the assertion is true. 
This is an opportunity for an induction proof. Assume the assertion is true up to 
$(f_{n-1}, f_{n})=1$. Then $(f_{n}, f_{n+1}) = (f_{n}, f_{n} + f_{n-1}) = (f_{n}, f_{n-1}) = 1.\;\;$ &#x2610;

#### 1.20 Use Euclid's remainder method to find $d = (826, 1890)$ and find $x, y \backepsilon d = 826x + 1890y$ 




$r_0 = 1890 \ \ \ \  r_1 = 826$


$1890 = 826 \cdot 2 + 238$


$826 = 238 \cdot 3 + 112$


$238 = 112 \cdot 2 + 14$


$112 = 14 \cdot 8 + 0$


$d = 14$


$1890 = 135 \cdot d$


$826 = 59 \cdot d$


$59 \cdot x + 135 \cdot y = 1$


The linear combination can be done by a tedious recitation of multiples, as in: 


Multiples of $135$: $135, 270, 405, 540, 675, 810, 945 = 135 \cdot 7, \dots$


Multiples of $59$: $59, 118, \dots, 944 = 59 \cdot 16$.


$14 = 826 \cdot (-16) + 1890 \cdot 7$


$x = -16, \; y = 7$.


A more elegant method takes the Euclid sequence in reverse, as Chip shows:


$14 = 238 - 112 \cdot 2$


Repeated substitutions...


$14 = (1890 - 2 \cdot 826) - 2 \cdot (826 - 3 \cdot (1890 - 2 \cdot 826))$


$14 = 1890 - 2 \cdot 826 - 2 \cdot 826 + 6 \cdot (1890 - 2 \cdot 826)$


$14 = 7 \cdot 1890 - 16 \cdot 826$ so again $x = -16$ and $y = 7$.



This is $14$ in terms of $826$ and $1890$: The reverse process leads back to a linear 
combination of the two starting numbers.
