### LaTeX patterns


Bobby Fischer once said, "Different people feel differently about LaTeX." Jupyter has an uneasy
relationship with LaTeX but there is generally some way of doing *at least adequate* typesetting
of the math covered here. This notebook is a poor man's cheat sheet.


- [An Actual LaTeX reference site 'Oeis'](https://oeis.org/wiki/List_of_LaTeX_mathematical_symbols).


Here we have Möbius and Erdös... how about a `ket`: $| \Psi \rangle$.
A `hat`: $\hat{x}$ A `vector`: $\tilde{x}$ or if you prefer $\underset{\sim}{a}$.


#### QED tombstones (unicode)


- Filled tombstone: &#x220e; outline: &#x1f78e;
- Big (thin-line) square: &#x2b1c; ($\neg$ Jupyter); medium square: &#x25fb; (ditto)
- My pref: ```$(c, d)=1.\;\;$ &#x2610;``` renders as $(c, d)=1.\;\;$ &#x2610;


#### Equation formatting


Inline LaTeX equations use a single dollar sign delimiter `$`: $x = y$...


...*versus* standalone equations from a double `$$` delimiter: These are centered and less cramped.  


$$x = y$$


For left-justified standalone use this: `$\begin{align} <equation LaTeX> \end{align}$`. 
One can use `equation` in place of `align` but that is again centered.


$\begin{align}
x = y \iff y = x 
\end{align}$


$\begin{align}
\textrm{Using stackrel:  }
a \stackrel{?}{=} b \stackrel{?}{\iff} x = y 
\textrm{  or with overset...  }
a \overset{?}{=} b \overset{?}{\iff} x = y 
\end{align}$


uses `stackrel{}{}` because `\questeq` is not supported.

#### Common Rarities, Arcane Mundanae


- **larger**: \\exists < \\Large{\\exists} <= \\LARGE{\\exists} : $\exists \;\; \Large{\exists} \;\; \LARGE{\exists}$
    - `\\relsize` is no dice in Jupyter at this time but here is bigness and largeness:
        - \\bigl( \\Bigl( \\Biggl( $\; \rightarrow \; \bigl( \Bigl( \Biggl($
        - \\large( \\Large( \\LARGE( $\; \rightarrow \; \large( \Large( \LARGE($
        - \\large[ \\Large[ \\LARGE[ $\; \rightarrow \; \large[ \Large[ \LARGE[$
        - \\sum {\\large \\sum} {\\Large \\sum} $\; \rightarrow \sum {\large \sum} {\Large \sum} {\Huge \sum}$
        
        
</br>

Auto-sizing braces: `$\left\{ ... \right\}$`:  $\left\{ \sum_{d|n} \frac{n}{d} \right\}$

$\begin{align}\left\{ \sum_{d|n} \frac{n}{d} \right\} \;\;\; \left\{ x \right\}\end{align}$

- **text inclusion via** `\\textrm{ can be found in }`: $x \textrm{ can be found in } y$
- **such that**: `\\ni` $\ni$ (bonus: Very Pythonic) or the more cramped approach `\\backepsilon` $\backepsilon$
- **element of**: \\in: $a \in B$
- **there exists**: \\exists $\exists$
- **for all**: \\forall $\forall x \in A$
- **logical negation**: \\neg $\neg a$
- **if and only if**: $\iff$
- **AND**, **OR**, **empty set**: \\cap and \\cup $\cap \;\; \cup$ and check out size modifiers:
$A \ \bigcap \ B  = \large{\emptyset}$ but I actually prefer $A \ \bigcap \ B  = \small{\emptyset}$.
- add **space**: \\; so that $x y$ becomes $x \; y$. Can also use backslash alone, followed by a space. $x \ y$
- **product** and **sum** notation: \\prod, \\sum: $\prod_{a}^{b} \;\;\; \sum_{c}^{d}$
- Greek **variants**: \\phi versus \\varphi, $\phi \textrm{ versus } \varphi$
- **Set archetypes**: Real numbers etcetera: \\mathbb{R}, \\Large{same} but \\R is a no-go: $\mathbb{R} \;\; \Large{\mathbb{R}}$
- **absolute value, divides etc**: Compare pipe character with \\mid and \\big|: $\;\; | a | \;\;\; \mid a \mid \;\;\; \big| a \big|$
- **norm** `\$\lVert object \rVert\$`  appears as  $\lVert object \rVert$
- **Big oh notation using \\mathcal** (calligraphy):
$\mathcal{O}(f(x))$ and to belabor: $\mathcal{A B C D E F G H I J K L M N O P}$
- Floor: use \\lfloor etcetera: $\lfloor x \rfloor$


Subscript pushed down a bit with `\lower1ex`; compare without on the right.


$\begin{align}
\sigma_{ \lower1ex{1 - \alpha}} - \sigma_{1 - \alpha}
\end{align}$


Raising and lowering operators...


$\begin{align}
\sigma  \lower1ex{\sigma} \lower2ex{\sigma} \lower4ex{\sigma} \sigma \raise3ex{\sigma} \sigma
\end{align}$

- **Definition, defined as:** $\;\; := \;\;$ or $\;\; \stackrel{\mathrm{def}}{=} \;\;$
- **Congruent** $\equiv$

### Multi-line constructs, variations in justification


Equation sequence, aligned on the \& character with \\ \\ for Linefeed: 


`$\begin{align}
\sigma(n) 
= \; & \sigma(2^{a-1}) \cdot \sigma(p) \\
= \; & (2^a-1) \cdot (p + 1) \\
\end{align}$`


$\begin{align}
\sigma(n) 
= \; & \sigma(2^{a-1}) \cdot \sigma(p) \\
= \; & (2^a-1) \cdot (p + 1) \\
= \; & (2^a-1) \cdot (2^a - 1 + 1) \\
= \; & 2 \cdot 2^{a-1} \cdot (2^a - 1) \\
= \; & 2n \\
\end{align}$

...or multiple right-justified equations in a sequence using \\ \\ as one LF...

$\begin{align}
a = 4 \\\\
b = 7 a \\
d = b - a - x \\
e = q + 4
\end{align}$

...or multiple left-justified equations...


$\begin{align}
& a = 4 \\\\
& b = 7 a \\
& d = b - a - x \\
& e = q + 4
\end{align}$

Multiple conditions


$\begin{align}
\mu(n) = \begin{cases} 
1 & \textrm{ if } n = 1 \\
-1^k & \textrm{ if } n \textrm{ is the product of } k \textrm{ distinct primes } \\
0 & \textrm{ otherwise }
\end{cases}
\end{align}$



<br><br><br>


Using \\atop to center-stack clauses


$\begin{align}
k > d \atop 
(k, d) = 1 \textrm{ and } k \textrm{ is even }
\end{align}$


This mechanism is not implicitly stackable to $n$ clauses; but for what it is worth it
can be nested (with diminishing font size):


$\begin{align}
k=1 \atop 
{
(k, d) = 1 \atop
\textrm{ and or perhaps other conditions such as } d = 19}
\end{align}$


...and of course we can get carried away as in the tale of Fury...


$\begin{align}
k=1 \atop
{(k, d) = 1 \atop
{\textrm{ I'll be judge I'll be jury }  \atop 
\textrm{ like maybe } d = 19
} }
\end{align}$

...so now magnify with \\Large:

$\begin{align}
\Large{
k=1 \atop
{(k, d) = 1 \atop
{\textrm{ here we find some additional conditions }  \atop 
\textrm{ like maybe } d = 19
} } }
\end{align}$



Finally here is the main use: \\atop in a sum (note '\\Biggl('):

            
$\begin{align}
(u * F^{*})(n) = 
\sum_{d|n} F^*(d) = 
\sum_{d|n} \sum_{k=1 \atop (k, d) = 1}^{d} f\Biggl( \frac{k}{d} \Biggr)
\end{align}$
