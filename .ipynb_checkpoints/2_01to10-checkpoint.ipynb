### Ground rules


We have three distinct forms of multiplication in chapter two.
First multiplication of two numbers as $2 \cdot 3 = 6$. Second implicit 
multiplication of composite functions: 
$a = f \cdot g \implies a(n) = fg(n) = f(n) \cdot g(n)$; 
and third Dirichlet convolution $f * g$ which is a sum of products 
over divisors of $n$.


Arithmetic functions are defined on the positive integers. 
Multiplicative functions are arithmetic functions for which 
$f(m \cdot n) = f(m) \cdot f(n)$ when $(m, n) = 1$. 
(I sometimes abbreviate 'f is multiplicative' as $f \Large{\times}$.)
Completely multiplicative functions (abbreviation: $f C \Large{\times}$) have 
$f(m \cdot n) = f(m) \cdot f(n) \; \forall \; m, n$.

More conventions:


- $p$ will indicate some prime


- $n$ is a generalized composite integer with $q$ distinct prime factors: $n = \prod_{i=1}^{q}{p_i}^{a_i}$


- $\nu(n)$ is the number of unique prime factors of $n$


- The radical of $n$ shorthand $rad(n)$ is the product of unique prime factors of $n$


- $I(n)$ is the identity function $\{1, 0, 0, 0, \dots \}$; also expressed as $\lfloor \frac{1}{n} \rfloor$.
Self evidently completely multiplicative.


- $\mu(n)$ is the Möbius function
  - $\mu(1) = 1$...
  - For $n > 1$: $\mu = 0$ when $n$ includes a squared prime...
  - Otherwise $\mu(n) = -1^{k}$ for $n$ having $\nu(n) = k$ distinct prime factors.
  - The generalized Möbius function $\mu_{k}(n)$ is defined just prior to problem 2.36.


- $u(n) = 1$ is the unit function: $1$ for all $n$. It is the Dirichlet inverse of $\mu(n)$. 
Self-evidently completely multiplicative.


- $\mu * u = I$: The sum of $\mu(d)$ over divisors of $n$ is zero when $n>1$.


- The Möbius inversion formula (MIF) in summation form:


$\begin{align}
\;\;\;\;\;\;\;\ f(n) = \sum_{d|n} g(d) = g*u \;\; \iff \;\; g(n) = \sum_{d|n} f(d) \cdot \mu \left( \frac{n}{d} \right) = f * \mu
\end{align}$


- Key idea: $u$ is the Dirichlet inverse of $\mu$ so the formula is symmetric: $f = g*u \iff f*\mu = g$.


- $\varphi(a)$ is the Euler totient, a count of relative primes less than or equal to $a$.
See the chapter 2 examples notebook for a relationship between $\varphi$ and the Discrete Fourier Transform.


- $\varphi$ and $\mu$ are $\Large{\times}$ but not $C \Large{\times}$.


- To reiterate: Multiplication appears in three forms
    - Dirichlet convolution $f * g$, a sum over products of $f$ and $g$
    - Composite functions: $\mu \cdot \varphi (n) = \mu(n) \cdot \varphi(n)$
    - Arithmetic multiplication in the usual sense where $6 \cdot 8 = 48$


In addition: $\Sigma$ sums are frequently transformed into $\Pi$ product forms.

<br> 


- The Dirichlet product of two multiplicative functions is itself multiplicative.


- $N(n) = n$ and more generally $N^k(n)$ is the power function equal to $n^k$: $k$ being a fixed real
or complex number. $N^k$ is completely multiplicative.


- $\varphi = \mu * N^1 = \mu * N$


- $d(n)$ is the number of positive divisors of $n$, multiplicative.
    - This is called the divisor function, a counting task 
    - $\sigma_{i}$ is the generalized divisor function
    - $d(n)$ is a special case of $\sigma_{i}(n)$
        - $\sigma_{i}(n)$ is the sum of divisors of $n$ each raised to the $i$.
        - $\sigma$ is multiplicative
        - $d(n) = \sigma_{0}(n)$
        - $\sigma_1(n)$ is useful in connecting perfect numbers to Mersenne primes; see 2.19

<br> 

- A common motif is 'index flipping'
    - Suppose we have a sum or product over divisors $d$ of $n$
    - The summand or multiplicand is function $f(d)$...
    - We can choose to flip the index without altering the resulting value of the sum or product
        - ...as in $\sum_{d|n} f(d) = \sum_{d|n} f(n/d)$
    - See 2.10 for a working example

### Recapitulating the proof of theorem 2.4


This proof contains an interesting view of product expansion in a combinatoric sense. 
Before starting on the theorem I quote the prior theorem 2.3 connecting the totient to 
the Möbius function:


$\begin{align}
\textrm{Theorem 2.3: }\;\;\; \varphi(n) = \sum_{d|n}\mu(d) \frac{n}{d} = n \sum_{d|n} \frac{\mu(d)}{d} = \mu * N
\end{align}$


Now to state **Theorem 2.4**: For $n \ge 1$


$\begin{align}
\varphi(n) = n \prod_{p|n} \left( {1 - \frac{1}{p}} \right)
\end{align}$.


That is: The totient can be written as its argument $n$ multipled by a product of ratios, 
one for each unique $p$ prime factor of $n$. Checking $n=13$ we have $13 \cdot (1-1/13) = 12$: 
so far so good. Checking $n=30$ we have prime factors $2, 3 \textrm{ and } 5$ so
$30 \cdot (1/2 \cdot 2/3 \cdot 4/5) = 8$; and listing the relative primes of $30$ less than $30$ 
we have $\{1, 7, 11, 13, 17, 19, 23, 29\}$ numbering 8 total, in agreement.


Proof: Strategy is take the right-hand $\prod$ expression and maneuver it to look like $\varphi(n)$.


First step: Ignoring the factor of $n$ until later, write the right-hand product in a 'counting all combinations of primes' fashion:

$\begin{align}
\prod_{p|n} \left( 1 - \frac{1}{p} \right) & = \left( 1 - \frac{1}{p_1} \right) \left( 1 - \frac{1}{p_2} \right) 
\cdots \left( 1 - \frac{1}{p_k} \right) \\
& = 1 - \frac{1}{p_1} - \frac{1}{p_2} - \cdots - \frac{1}{p_k} + \frac{1}{p_1 p_2} + \frac{1}{p_1 p_3} + \cdots + \frac{1}{p_{k-1} p_k}
- \frac{1}{p_1 p_2 p_3} - \cdots + \frac{{-1}^k}{p_1 p_2 \cdots p_k}
\end{align}$


The alternating signs in this sequence of fractions mirror the Möbius function: 
Each numerator is $\mu(\textrm{denominator})$; and powers of 2 or more for these
prime factors do not appear.


Rewriting with abstract indices to indicate 'all possible combinations': 


$\begin{align}
\prod_{p|n} \left( {1 - \frac{1}{p} } \right) & =
1 - \sum \frac{1}{p_i} + \sum \frac{1}{p_i \cdot p_j} - \sum \frac{1}{p_i \cdot p_j \cdot p_k} 
+ \cdots + \frac{{(-1)}^k}{p_1 \cdot p_2 \cdots p_k}
\end{align}$


Here is the Tommy-esque plot twist: Add additional terms that we know in advance sum
to nothing. This will take advantage of $\mu(n) = 0$ when $n$ contains a square.


In the above equation, the product expansion becomes a sum of fractions involving all possible 
combinations of prime factors of $n$. These various combinations all reside in fraction 
denominators. Suppose with Dirichlet convolution in mind that we want this sum to extend 
over *all* divisors $d$ of $n$, not just the primes raised to the first power. Such an extension
is possible provided the additional fractions involving primes raised to the power $2$ or
more contribute nothing to the sum. To guarantee this we place a Möbius function in 
the numerator of *all* the fractions: The existing ones from the equation above and the new 
fractions we are introducing. This Möbius function serves two purposes: It reduces each 
'square-ful denominator' fraction to zero; and for the other fractions it produces the 
necessary combinatoric sign changes. 


With the Möbius function in hand the sum can be expanded to a sum over all divisors $d|n$: 


$\begin{align}
\prod_{p|n} \left( {1 - \frac{1}{p} } \right) & = \sum_{d|n} \frac{\mu(d)}{d} \;
\end{align}$

Multiply by $n$ and make use of the result of theorem 2.3: The totient is the Dirichlet product of the Möbius function 
and the identity function $N(n) = n$:


$\begin{align}
n \prod_{p|n} \left( {1 - \frac{1}{p} } \right) = n \sum_{d|n} \frac{\mu(d)}{d} = \sum_{d|n} \mu(d) \frac{n}{d} = \mu * N = \varphi(n) \;\;\;\;\;\;\;
\end{align}$&#x2610;


Commencing with problems and solutions; as always with a tip of the hat to Chip Hurst. 

### 2.1 Find $n$ for which (a) $\varphi(n) = \frac{n}{2}$; (b) $\varphi(n) = \varphi(2n)$; (c) $\varphi(n) = 12$.


#### **2.1 (a) $\varphi(n) = \frac{n}{2}$** 


Euler's totient function $\varphi(n)$ yields the number of relative primes less than or equal to $n$. This is
a multiplicative arithmetic function, meaning $\varphi(m \cdot n) = \varphi(m) \cdot \varphi(n) \textrm{ when } (m, n) = 1$.


For $n=1$ we have $\varphi(1) = 1$ (not a solution) and $\varphi(2) = 1$ (a first solution).
These are the only two odd values of $\varphi(n)$. For all $n > 2$ we have $\varphi(n)$ is even.
This means it can be written as $2$ times some integer $\delta$ (which can itself be even or odd).


We have *a* solution $n=2$ so now to find *other* $n > 2$ solutions such that $\varphi(n) = 2 \cdot \delta = n / 2$. 
Notice this implies $n = 4 \cdot \delta$ so solution $n$ values will be multiples of $4$. Possibly $n$ contains
more powers of $2$ as well as some odd factors. Suppose then that for a solution $n$ there is an $\alpha > 1$ such
that $n = 2^{\alpha} \cdot \Pi \; \textrm{ where } \Pi \textrm{ is a product of some odd factors}$.

For formality let's rewrite $n$ using product notation for the various odd prime factors raised to their 
respective powers:


$\begin{align}n = 2^{\alpha} \cdot \Pi = 2^{\alpha} \cdot \prod_{p \; odd}{p_i}^{a_i} \end{align}$


Observe that $(2^\alpha, \; \Pi) = 1$ so that (since $\varphi(n)$ is multiplicative) we can break apart $\varphi(n)$:


$\begin{align}\varphi(n) = \varphi(2^\alpha \cdot \Pi) = \varphi(2^\alpha) \cdot \varphi(\Pi) \end{align}$


The first term $\varphi(2^\alpha)$ is equal to the count of odd numbers less than $2^\alpha$ which is $2^{\alpha - 1}$. All the even
numbers less than or equal to $n$ are *not* relatively prime to $n$ because they all contain a factor of $2$. So 
for solution values of $n$ we have:


$\begin{align} \varphi(n) = 2^{\alpha-1} \cdot \varphi(\Pi) = \frac{n}{2} = \frac{2^\alpha \cdot \Pi}{2} = 2^{\alpha - 1} \cdot \Pi\end{align}$

For this to be true we must have $\varphi(\Pi) = \Pi$. However for $\Pi > 2$ we know $\varphi(\Pi)$ is even.
To avoid this contradiction we conclude that $\Pi$ must be $1$: No odd factors in $n$.


Consequently $n = 2^\alpha = \{ 2, 4, 8, \dots \}$.

#### **2.1 (b) $\varphi(n) = \varphi(2n)$** 


Useful observation: $\varphi(2^s) \ne \varphi(2^{s+1})$ excepting the one case $s=0$.


Write $n$ as a product of (a power of 2) times (a residual odd number $\Pi$).
Here $\Pi$ is **either** $1$ **or** a residual product of odd primes raised to
various powers). We can write $n = 2^s \cdot \Pi$, just as in part (a).


Write $2n$ similarly: $2n = 2^{s+1} \cdot \Pi$


Use the multiplicative nature of $\varphi$ to state the condition on $n$: 


$\begin{align}\varphi(2^s) \cdot \varphi(\Pi) = 
\varphi(2^{s+1}) \cdot \varphi(\Pi)\end{align}$


This will be true for *any* odd $\Pi$ so long as $s$ has value $0$
so the solution is $n = \{1, 3, 5, 7, 9, \dots\}$. For these values of $n$ we have $\varphi(1) = \varphi(2), \; \varphi(3) = \varphi(6), \; \varphi(5) = \varphi(10)$ and so on.

#### **2.1 (c) $\varphi(n)=12$**


It is easy enough to guess that $n=13$ is a solution since all the numbers less than
a prime $p$ are relatively prime to $p$; so $\varphi(p) = p - 1$. However I found it
arduous to conclusively produce all other possible solutions; here goes!


Let's begin with the *product* form of the totient, in contrast with a counting sum from the definition:


$\begin{align}
\varphi(n) = 12 = 2 \cdot 2 \cdot 3 = n \prod_{p|n} \left( 1 - \frac{1}{p} \right) = n \prod_{p|n} \frac{p_i-1}{p_i} \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(1)
\end{align}$
 


using theorem 2.4 (see above). Quick check: $\varphi(13) = 12$ as expected. Now in the right 
expression of (1) the entirety of the denominator product cancels factors of the leading $n$,
leaving a residual $R \ge 1$: 


$\begin{align}12 = R \cdot \prod_{p|n}(p_i-1)
\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(2)
\end{align}$



Now for steps to equation (3): From equation 1 write $n$ as a product of primes raised to powers.


$\begin{align}n = \prod_{p|n}{p_i}^{\alpha_i}\end{align}$.


Match up these $p_{i}$ factors with the $(1 - \frac{1}{p_i})$ terms in the righthand side product of equation (1):


$\begin{align}p_i^{\alpha_i} \cdot \left( 1-\frac{1}{p_i} \right) = p_i^{\alpha_i} - p_i^{\alpha_i-1}\end{align}$.


This arrives at a new expression for the totient:


$\begin{align}
\varphi(n) = 12 = 2 \cdot 2 \cdot 3 = \prod_{p|n} \bigl( p_i^{\alpha_i} - p_i^{\alpha_i - 1} \bigr)
\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(3)
\end{align}$


which also works as expected for $n=13$. Now we have a basis for breaking the problem
down into cases, specifically "how many distinct prime factors does $n$ have?" where
$\varphi(n) = 12$. 


Suppose $n$ has $k$ distinct prime factors. If $k \ge 4$ then equation (1) implies
$12$ has at least 3 even factors; which it does not. Consequently $k$ must be 1, 2 or 3.


Detail: For $k \ge 4$ equation (2) will include 4 or more
primes, one of which might be 2. The remaining 3 or more primes will all be odd; and the 
expression $(p_i - 1)$ will yield a product of 3 or more even numbers. This is how
$k \ge 4$ is eliminated leaving the cases $k = 1, \; k = 2, \; \textrm{ and } k = 3$.



For $k \le 3$ we need *up to* three primes so we call them $p, \; q, \; \textrm{and} \; r$.
Further they are raised to powers $\alpha, \; \beta, \; \textrm{and} \; \gamma$.
So for a $k=3$ solution we would have $n = p^\alpha \cdot q^\beta \cdot r^\gamma$.
Lets also order these primes: $p < q < r$. 
Each distinct prime contributes a factor to the product of equation 3. 


For $k=1$ we have $12 = p^\alpha - p^{\alpha-1}$. For $k=2$ we have 
$12 = (p^\alpha - p^{\alpha-1})\cdot(q^\beta - q^{\beta-1})$. For
$k=3$ this extends once more to 
$12 = (p^\alpha - p^{\alpha-1})\cdot(q^\beta - q^{\beta-1})\cdot(r^\gamma - r^{\gamma-1})$.


These constraints suggest a table to find all $n$ for which $\varphi(n) = 12$. That is,
the columns of the table are for successive primes and the rows are for successive
exponents for the primes. The values in the table for a particular $p$ and $\alpha$ 
are $\bigl( p^{\alpha} - p^{\alpha - 1} \bigr)$. Here is such a table:

<table style="margin-left: 0;">
<tr><th></th><th>p:</th><th>2</th><th>3</th><th>5</th><th>7</th><th>11</th><th>13</th></tr>
<tr><td>$\alpha = $</td><td><b>1</b></td><td>1</td>
    <td>2</td><td>4</td><td>6</td><td>$\cdot$</td><td>12</td></tr>
<tr><td>$\alpha = $</td><td><b>2</b></td><td>2</td><td>6</td>
    <td>$\cdot$</td><td>$\cdot$</td><td>$\cdot$</td><td>$\cdot$</td></tr>
<tr><td>$\alpha = $</td><td><b>3</b></td><td>4</td><td>$\cdot$</td>
    <td>$\cdot$</td><td>$\cdot$</td><td>$\cdot$</td><td>$\cdot$</td></tr>
</table>

Table values must be multiplied together to produce $12$. So for example we have $p=13$ and $\alpha=1$ 
(with $k = 1$) producing $12$; so the solution is $n=13$ as noted above. Of course $12 \cdot 1 = 12$
as well, meaning we can take $k = 2$ and use $p = 2$ with $\alpha = 1$ and $q = 13$ with $\beta = 1$.
In this case $n = 2 \cdot 13 = 26$, a second solution. (This echoes the result of (b) above.)


Since we used up the only $1$ table value, the $12$ at $p = 13, \; \alpha = 1$ is "used up" for possible
solutions. Primes larger than $13$ will have $\varphi(p) > 12$ so the remaining solutions will all
originate from the lower primes $2, \; 4, \; 5, \; \textrm{and } 7$. For these $k$ must be $2$ or $3$.


Suppose $k=2$: The strategy is to choose pairs of numbers from the above table
whose product is $12$ and from these pairs infer $p$, $\alpha$, $q$ and $\beta$.
We can begin with the minimum values $p=2$ and $\alpha = 1$ to get a 
$p$-term of $1$ in the product. Chosing $q=13$ and $\beta = 1$ gives the desired 
totient $1 \cdot 12 = 12$; so $n=2 \cdot 13 = 26$ is a solution. This is consistent with 
$\varphi(2) = 1$ and $\varphi$ is multiplicative. 


All of the table pairs for $k=2$ are easily read off: 

- $(1 \cdot 12)$ corresponding to $2^1 \cdot 13^1 = 26$
- $(2 \cdot 6)$ corresponding to $2^2 \cdot 3^2 = 36$
- $(2 \cdot 6)$ corresponding to $2^2 \cdot 7^1 = 28$
- $(2 \cdot 6)$ corresponding to $3^1 \cdot 7^1 = 21$


Finally there is one $k=3$ combination: $1 \cdot 2 \cdot 6$ corresponding to $2^1 \cdot 3^1 \cdot 7^1 = 42$.
The exhaustive set of values for which $\varphi(n) = 12$ is consequently $\{13, 26, 36, 28, 21, 42 \}$.

### 2.2 Prove or find counterexamples $\otimes$ for...


(a) Show $(m,n) = 1 \implies (\varphi(m), \varphi(n))=1$. 
Choosing $m=3,\;n=4$ both give $\varphi = 2$, a counterexample. $\otimes$


(b) Show $n \; composite \; \implies (n, \varphi(n)) > 1$.
Choose $n = 15$ and check: $(15, 8) = 1$, a counterexample. $\otimes$


(c) If the same primes divide $m$ and $n$ then $n\cdot\varphi(m) = m\cdot\varphi(n)$.


$\frac{\varphi(m)}{m}=\prod_{p|m}{1-p^{-1}}$. Since all $p$ that divide $m$ also divide $n$
the product on the right is also equal to $\frac{\varphi(n)}{n}$.


This part (c) is very concise so I will expand a little bit. Writing the product form of 
the totient function $\varphi(s)$ introduces a factor of $s$; so divide both sides
by $s$ to get $\varphi(s)/s$ on one side and the product on the other. Here
this product will be identical for two distinct numbers $m$ and $n$ that share the same
set of prime factors; so their totient-to-self ratios are equal.  $\;$ &#x2610;

### 2.3 Show that $\frac{n}{\varphi(n)}=\sum_{d|n}\frac{\mu^2(d)}{\varphi(d)}$


## Re-do this rather lost write-up


From Theorem 2.15 we have a rule of multiplicative functions: That $f * g$ is multiplicative if 
both $f$ and $g$ are multiplicative. Since the problem includes a Dirichlet-like sum 
$\sum_{d|n}\frac{\mu^2(d)}{\varphi(d)}$, an idea is to introduce $u(n)=1$ to arrive at 
$\frac{\mu^2}{\varphi}*u$.


The composite function $\frac{\mu^2}{\varphi}$ is constructed from multiplicative 
functions $\mu$ and $\varphi$. However: Missing is a demonstration that this 
construction is multiplicative. 

Prime decomposition of $n$ in terms of $k$ distinct primes $p_i$ with corresponding 
exponents $a_i > 0$.


$
\begin{align}
{
\sum_{d|n}\frac{\mu^2(d)}{\varphi(d)} = 
\frac{\mu^2(n)}{\varphi(n)} * u(n) = 
\prod_{i=1}^k \Bigl( \sum_{j=0}^{a_i} 
\frac{\mu^2(p_{i}^{j})}{\varphi(p_{i}^{j})} \Bigr)
}
\end{align}
$




From here we can make use of features of $\mu$ and $\varphi$.
First *inside the sum* the function $\mu^2(q)$ has value 
$1$ when $q$ is either $1$ or a prime, 
and zero otherwise. Second, $\varphi(1)=1$ and $\varphi(p)=p-1$.
As a result the summation over powers of the prime factor $p_i$
simplifies into just the $j=0$ and $j=1$ cases:


$
\begin{align}
{
\sum_{d|n}\frac{\mu^2(d)}{\varphi(d)} = 
\prod_{i=1}^k \Bigl( 1 + \frac{1}{p_i-1} \Bigr) = 
\prod_{i=1}^k \Bigl( \frac{p_i}{p_i-1} \Bigr) = 
\prod_{p|n} \frac{1}{1-p^{-1}} =
\frac{n}{\varphi(n)}.
}
\end{align}
$ 


The last step is from the product form of the totient function; and there we are. $\;$ &#x2610;

### Intermezzo on $1$ and *multiplicative functions*


The chapter problems occasionally leverage some version of "multiply by 1". In the arithmetic 
function menagerie we find $1$ in both the identity function $I(n) = \bigl[ \frac{1}{n} \bigr]$ 
and in the unit function $u(n)=1$. For example in 2.3 the unit function with no dependency on its
argument was used to set up a convolution with $\frac{\mu^2}{\varphi}$.


Another motif is the sum $\sum_{d|n}$ appearing with some arguments translating to a Dirichlet 
convolution of multiplicative functions. This becomes a decomposition of $d|n$ into products of sums 
of powers of primes with $(p_a, p_b)=1$. That is, a function being multiplicative gives us a 
decomposition of functions of prime products as products of functions of primes.

### 2.4 Show that $\varphi(n) > \frac{n}{6} \; \forall \; n $ with $k \le 8$ prime factors.


This is proof by calculation.


Taking $\varphi(n) = n \cdot \prod_{p|n}\frac{p-1}{p}$ the goal is to show the
product $\prod < \frac{1}{6}$. Calculate $\prod$ for
the first eight primes: 
$\{\frac{1}{2} \cdot \frac{2}{3} \cdot \frac{4}{5} \cdot \cdots \cdot \frac{18}{19}\}$.
The result is greater than $1/6$. Choosing a different set of eight
primes produces larger factors (closer to 1) so their product is also greater
than $1/6$. Likewise choosing fewer than eight unique prime factors results in a larger 
product. In both cases, therefore, the limiting product above 
gives a lower bound greater than $1/6.\;\;$ &#x2610;

### Remark on alternating-sign binomial coefficient sums


Concerning sums of binomial coefficients: I will index elements of the Pascal triangle $P$
from $0$ counting rows downward and elements across each row. 


```
Row
 0               1
 1             1   1
 2           1   2   1
 3         1   3   3   1
 4       1   4   6   4   1
 5     1   5  10   10  5   1
          ... etcetera ...
```

Row $n$ contains $n+1$ elements; for example row $4$ is $1, 4, 6, 4, 1$. 
These elements are indexed $j = 0, 1, 2, 3, 4$. 


Definition: Row $i$ element $j$ of $P$ has value $\binom{i}{j}$.


Produce a triangle $A$ by applying to $P$ an alternating-sign multiplier along each 
row.


```
Row
 0               1
 1             1  -1
 2           1  -2   1
 3         1  -3   3  -1
 4       1  -4   6  -4   1
 5     1  -5  10  -10  5  -1
          ... etcetera ...
```

Now row $0$ is $1$ as before; but row $4$ has become $1, -4, 6, -4, 1$.



As a formula the $j$'th element of the $i'th$ row has value $A_{ij} = {-1}^{(i+j)} \cdot \binom{i}{j}$.



Claim: Row $0$ of $A$ sums to $1$ and row $n>0$ sums to $0$. (Not even Tommy
can stop me from making this claim.)



Define a weighted Pascal triangle $P_1$ as follows: Multiply each element of
$P$ by its row index. Row $0$ of $P_1$ is still $1$ but now row $1$ is $0, 1$. 
Row $4$ of $P_1$ is $0, 4, 12, 12, 4$, row $5$ is $0, 5, 20, 30, 20, 5$.
Now apply the same alternating sign multiplier as before to produce $A_1$.


Claim: Row $0$ of $A_1$ sums to $1$ and row $n>0$ of $A_1$ sums to $0$.

### 2.5 Show $f = \mu * \nu$ is 0 or 1 


Reminder: $\nu(n)$ is the number of distinct prime factors of $n$. 


This is a good problem threading the non-trivial versus tractable needle. It annotates Chip's 
solution and fixes a typo.


***Note: Problem 2.5 appears in the chapter 2 examples notebook.***


Plan: Lay out useful tools; then work through five cases for $n$: 
(1) $n = 1$, (2) $n = p$, (3) $n = p^a$ for $a > 1$, (4) $n = \prod p_i$ 
for 2 or more primes (i.e. $n$ is squarefree), 
and finally (5) $n = p^a \cdot m$ with $m > 1$.

- $p$ is a prime unless otherwise stated; whereas $q$ may or may not be prime
- $\nu(1)=0$ so $\nu$ is not multiplicative
- For $a \ge 1$ and supposing $(p, q) = 1$ we have $\nu(p^a \cdot q) = 1 + \nu(q)$.
- $\mu$ is multiplicative so $\mu(p \cdot q) = \mu(p) \cdot \mu(q) = -\mu(q)$.

Case 1, $n = 1$: We have $f(1) = 0$, check. 

Case 2, $n = p$: $f(p) = 1 \cdot 1 + -1 \cdot 0 = 1$, check.

Case 3, $n = p^a$ where $a > 1$: $f(p^a) = 1*1 + -1*1 + 0*1 + \dots + 0*1 + 0*0 = 0$, check.

Case 4, $n > 1$ is composite and square-free: $n=\prod p_i$.


(Case 4 is a specialization of Case 5.)


Here we separate say the first prime factor of $n$--call this $p$--such that $p \cdot m = n$
where of course $(p, m) = 1$ and $m > 1$. First: Write out the Dirichlet sum and then 
partition it into two sums over non-overlapping divisor sets:
{divisors of $n$ that *do not* contain $p$} and {divisors of $n$ that *do* contain $p$}.

$\begin{align}
f(n) = \sum_{d|n} \mu(d) \cdot \nu \left( \frac{n}{d} \right) =
\sum_{d|m} \mu(d) \cdot \nu \left( \frac{n}{d} \right) + \sum_{(p \cdot d)|n} \mu(p \cdot d) 
\cdot \nu \left( \frac{n}{p \cdot d} \right)
\end{align}$

In the expanded expression on the right: The left sum $\nu$ argument 
$\frac{n}{d}$ gets substitution $n = p \cdot m$. Meanwhile the right sum factor 
divisor condition $(p \cdot d) | n$ becomes $d | m$. This does not change
the $\mu$ and $\nu$ function arguments: If $(p \cdot d) | n$ then 
$d | m$ since $n = p \cdot m$ with $p$ and $m$ relatively prime. 


As an example take $p=2$ and $n=210$ so $m=105$. In the above right-hand sum we have $(pd)|n$ so
the $d$ values are $\{ 1, 3, 5, 7, 15, 21, 35, 105 \}$. In the substitution to follow below we
have $d|m$: The same $d$ values. We factor out $p$ from the sum
condition without changing the summands.


Lastly there is also a substition of $m$ for $\frac{n}{p}$ in the 
right-hand $\nu$ argument. This gets us the following expression:

$\begin{align}
f(n) = \sum_{d|m} \mu(d) \cdot \nu \left( p \cdot \frac{m}{d} \right) + \sum_{d|m} \mu(p \cdot d) 
\cdot \nu \left( \frac{m}{d} \right)
\end{align}$

Then follows some application of the tools noted above.


$\begin{align}
f(n) = \sum_{d|m} \mu(d) \cdot \left( 1 + \nu \left( \frac{m}{d} \right) \right) 
+ 
\sum_{d|m} -\mu(d) \cdot \nu \left( \frac{m}{d} \right)
\end{align}$

Now to distribute the left sum interior product:


$\begin{align}
f(n) = \sum_{d|m} \mu(d) 
+ \sum_{d|m} \mu(d) \cdot \nu \left( \frac{m}{d} \right) 
- \sum_{d|m} \mu(d) \cdot \nu \left( \frac{m}{d} \right)
\end{align}$


$\begin{align}
f(n) = \sum_{d|m} \mu(d) = 0 \textrm{ as } m > 1. 
\end{align}$ 


Case 4: Check.


Case 5 covers the remainder of the integers: $p$ is prime and $a$ is greater than one; and
we can set $n = p^a \cdot m$ (noting $(p, m) = 1$).
Here $m > 1$ and $m$ may or may not include a squared prime factor.


Here we break out the Dirichlet sum over divisors of $n$ in terms of powers of $p$:


$\begin{align}
f(n) = \mu(n) * \nu(n) = \sum_{d|n} \mu(d) \cdot \nu \left( \frac{n}{d} \right) = 
\sum_{i=0}^{a} \sum_{\substack{p^i \cdot d|n \\ (p, d) = 1}} 
\mu (p^i \cdot d) \cdot \nu \left( \frac{n}{p^i \cdot d} \right) 
\end{align}$

The second condition $(p, d) = 1$ ensures we do not double-count factors. 


(Aside: If Case 5 covered Case 4 this solution would be much simpler.)


Ok now *quo vadis*? First of all -- in a brilliant move -- the outer sum need not
go beyond $i = 1$ because in such circumstances the argument of $\mu$ contains a square. 

Secondly the inner sum uses the same change of divisor conditions as before, again
because $(p, m) = 1$: Rather than $p^i d | n$ we can use $d | m$. 


Thirdly the numerator of the $\nu$ argument is $n = p^a \cdot m$.
This combines with the $p^i$ in the denominator. 


$\begin{align}
f(n) = \sum_{i=0}^{1} \sum_{d|m} \mu(p^i \cdot d) \cdot \nu \left( p^{a-i} \cdot \frac{m}{d} \right)
\end{align}$

Now we eliminate the outer $i$-sum by adding the two cases: $i = 0$ and $i = 1$.
In the second sum: Also apply the $\mu(p \cdot d) = -\mu(d)$ mechanism.

$\begin{align}
f(n) = \sum_{d|m} \mu(d) \cdot \nu \left( p^{a} \cdot \frac{m}{d} \right)
+
\sum_{d|m} -\mu(d) \cdot \nu \left( p^{a-1} \cdot \frac{m}{d} \right)
\end{align}$

Home stretch: Apply the $\nu(p^a \cdot m) = 1 + \nu(m)$ mechanism. This requires $a > 1$
to avoid a $p^0$ situation.


$\begin{align}
f(n) = \sum_{d|m} \mu(d) \cdot \left( 1 + \nu \left( \frac{m}{d} \right) \right)
- \sum_{d|m} \mu(d) \cdot \left( 1 + \nu \left( \frac{m}{d} \right) \right)
= 0
\end{align}$


Result: $f(p) = 1$ and $f = 0$ otherwise. $\;\;$ &#x2610;


Continuing the aside from above: It seems that running through case 5 
with $a$ set to $1$ would follow the same path until the last step 
when $p^{a-1}$ is just $1$ so the right-hand $\nu$ is a bit different: 


$\begin{align}
f(n) = \sum_{d|m} \mu(d) \cdot \left( 1 + \nu \left( \frac{m}{d} \right) \right)
- \sum_{d|m} \mu(d) \cdot \left( \nu \left( \frac{m}{d} \right) \right)
\end{align}$


$\begin{align}
f(n) = \sum_{d|m} \mu(d) = 0
\end{align}$


Case 5: Check.

### 2.7 Divisor sum of a Möbius product


Notation: $\mu((p, d))$ is $\mu(GCD(p,d))$. 


Confirm given some $n$ and some prime $p$ that the sum


$\begin{align}
\sum_{d|n}\mu(d)\cdot\mu((p,d))
\end{align}$


...is equal to


- $1$ if $n=1$
- $2$ if $n=p^\alpha$
- $0$ otherwise


Case 1: $n=1$: $d=1$, and the sum becomes $\mu(1) \cdot \mu(1) = 1$.

Case 2: $n=p^\alpha$. 


For $s$ not squarefree $\mu(s) = 0$. Only two terms remain in the sum: $d=1$ and $d=p$.


$\begin{align}
\sum = \mu(1) \cdot \mu((p, 1)) + \mu(p) \cdot \mu((p,p)) = 1\cdot 1 + (-1) \cdot (-1) = 2, \textrm{ check}
\end{align}$

Case 3 we can break downto two sub-cases: $n$ does not / does have $p$ as a factor.


Case 3a: $n$ does not have $p$ as a prime factor: $n = \prod q_i^{\beta_i}$.


Now terms in the sum over divisors of $n$ are non-zero only when $d$ is a product
of first powers of primes $\{ q_i \}$. The Möbius function will yield powers of 
$-1$ for the various $\mu(d)$; and $\mu((p, d))$ will always be $$\mu(1) = 1$. 
The resulting sum will be that of a sequence of alternating sign binomial 
coefficients (see 2.5 above) which is zero. This concludes Case 3a. 


Case 3b: $n$ *does* include $p$ as a prime factor (say to some power $\alpha$). 


As in Case 3a only squarefree divisors of $n$ will contribute to the sum. Divisors of $n$ 
that do not include $p$ are covered by Case 3a which sums to $0$; but this is not the
complete set of divisors $d$ in the sum. We also need $d$ that include $p^1$ as a factor. 
(Möbius again: $p^2 \dots p^\alpha$ all contribute $0$.)


Case 3a divisors are now replicated with an added factor of $p$. This gives the same 
alternating sign binomial coefficient sum multiplied by (-1) from the added factor $p$.
The second factor will always be $\mu((p,d)) = \mu(p) = -1$ and this cancels the 
factor of $-1$ in $\mu(d)$. That is, we have the same sum as in Case 3a, again $0$,
with a total sum of $0$ completing 3b and the entire third case. $\;\;$ &#x2610;

### 2.8 Show $\sum_{d|n}\mu(d) \cdot \log^m d = 0$ if $m>0$ and $n$ has $\nu(n) > m$


Might "sum over $d$ of $\mu(d)$ equals 0" be of use here?


Chip's solution features clever manipulation of sums and an induction strategy. This 
assumes the sum is $0$ for values of $m: \;1, \; 2, \; \dots \textrm{ some k } < \nu(n)$; 
then shows the sum is also zero for $m = k + 1 < \nu(n)$.  With this done it remains to 
check the ground floor: $\sum = 0$ for $m=1$.

Presume $n$ is square-free. Why? Because a square-ful $n$ has square-ful divisors $d$ that
all contribute $0$. So: Take $n = \prod_{i=1}^{\nu(n)} p_i$.


For $m=1$ the sum should be $0$. Let's test that with $n = 30 = 2 \cdot 3 \cdot 5$ giving 
$d =  1, 2, 3, 5, 6, 10, 15, 30$. $\log 1 = 0$ so consider the remaining primes and composites. 
For each $d$ I convert $\mu(d)$ to $\pm 1$ and expand $\log (a\cdot b)$ as $\log a + \log b$. 


Result: $-\log 2 - \log 3 - \log 5 + \log 2 + \log 3 + \log 2 + \log 5 + \log 3 + \log 5 - \log 2 - \log 3 - \log 5 = 0$.


That is, the expansion of the sum of Möbius-logarithm products introduces an equal number 
of $+$ and $-$ signs for each prime factor. Let's do simpler checks with $n=6$ and $n=11$:

$\begin{align}\sum_{d = 1, 2, 3, 6} \mu(d) \cdot \log {d} = -\log 2 - \log 3 + \log 2 + \log 3 = 0.\end{align}$

And now the rule breaks down for $n = 11$ as $m = 1 = \nu(11)$.


$\begin{align}\sum_{d = 1, 11} \mu(d) \cdot \log {d} = -\log 11 \ne 0\end{align}$

That's not really rigorous but the examples are consistent with 'The sum is zero for $m=1$.' 
Now per induction assume this sum is zero for $m = 1, 2, \dots$ up to some $k$. Now we sort
out the sum for $m = k + 1$ still with $\nu(n) > k + 1$. 


$\begin{align}{\sum_{d|n} \mu(d) \log^{k+1}d = \sum_{d|n} \mu(d) \cdot \log^k d \cdot \log d}\end{align}$.


Factoring the logarithm raised to $m + 1$ as two logarithms is the first of a sequence of inspired 
maneuvers. To follow: $\log d$ is reconstrued as the sum of logs of prime factors of $d$. For this 
step let's skip the $d = 1$ case as $\log 1 = 0$.


$\begin{align}{\sum_{d|n} \mu(d) \log^{k+1}d = \sum_{d|n} \mu(d) \cdot \log^k d \cdot \sum_{p_i|d}\log p_i}\end{align}$


Retaining a concrete example might help so let's bump up from $n = 30$ to $n = 210$ with four
distinct prime factors $2, 3, 5, 7$. This puts $k$ at 1 or 2 but lets keep $k$ as abstract. 
Now the sum over $d|n$ covers $\{ 1, 2, 3, 5, 7, 6, 10, 14, 15, 21, 35, 30, 42, 105, 210 \}$
although again $d = 1$ contributes $0$. 


The next step requires a good deal of staring: It reverses the order of the sums and this comes with
a change in both sum conditions.


$\begin{align}{\sum_{d|n} \mu(d) \log^{k+1}d = \sum_{d|n} \mu(d) \cdot \log^k d \cdot \sum_{p_i|d}\log p_i = 
\sum_{p_i|n} \sum_{\substack{d|n \\ p_i|d}} \log p_i \cdot \mu(d) \cdot \log^k d}\end{align}$



This remarkable transformation *requires* the revised conditions to produce all the same terms. 
It is sort of a restacking of those terms by a shift in emphasis. Consider that one of the terms 
in the upper sum (still for $n = 210$) is $\mu(10) \cdot \log^m(10) \cdot (\log 2 + \log 5)$.
Multiply this out to get two terms. 


Now with the sum order reversed the new outer sum produces these same two terms. 
This will be when $p_1 = 2$ and when $p_3 = 5$: From that outer sum over $p_i|n$. 
For each $p_i$ the inner sum goes through all the $d|n$ but with the additional
condition to only consider $d$ divisible by that outer sum prime: $p_i|d$. 
In the example we have $p_1 = 2$ giving us a term $\log 2 \cdot (\mu(10) \cdot \log^m(10))$.
Later on with $p_3 = 5$ we get the second term $\log 5 \cdot (\mu(10) \cdot \log^m(10))$.


Notice in the above expression that $\log p_i$ depends only $p_i$ (not on $d$) so
it can be factored out of the inner sum and now we have


$\begin{align}{\sum_{d|n} \mu(d) \log^{m+1}d = \sum_{p_i|n} \log p_i 
\sum_{\substack{d|n \\ p_i|d}} \mu(d) \cdot \log^m d}\end{align}$


The next step is another remarkable one. The goal is to rewrite the inner sum above. 
I'm going to illustrate the strategy using the example numbers first and then describe
the transformation. 


Take the case of $p_1=2$ and now the inner sum will use $d = \{ 2, 6, 10, 14, 30, 42, 70, 210 \}$ so
there are eight values for $d$, all divisible by $2$. These will appear in the inner sum, as in
$\mu(2) \cdot \log^m(2) + \mu(6) \cdot \log^m(6) + \mu(10) \cdot \log^m(10) + \cdots + \mu(210) \cdot \log^m(210)$.
We could just as easily make the inner sum be over those same numbers divided by the outer 
sum prime $2$ provided that the function is modified to multiply back by the outer sum prime. 
That is: Dividing by $2$ we have a new set of divisors of $n/p_1$ and a 'restoring' version
of the Möbius and logarithm functions. The new $d = \{ 1, 3, 5, 7, 15,21, 35, 105 \}$ and the
new function is now $\mu(d \cdot p_i) \cdot \log^m d \cdot p_i$.


The net result here is that the inner sum produces all the same terms as before; but the 
outer sum prime $p_i$ is introduced in the inner sum as a factor of the Möbius and logarithm
arguments. 



$\begin{align}{\sum_{d|n} \mu(d) \log^{m+1}d = \sum_{p_i|n} \log p_i 
\sum_{d|\frac{n}{p_i}} \mu(d \cdot p_i) \cdot \log^m d \cdot p_i}\end{align}$


So far so good; and next: The argument of the Möbius function is a product of unique prime factors; 
so the $p_i$ factor contributes a minus sign to the end result. We can remove this $p_i$ and put a
negative sign out front. Also let's expand the logarithm of the product in the usual manner.


$\begin{align}{\sum_{d|n} \mu(d) \log^{m+1}d = -\sum_{p_i|n} \log p_i \sum_{d|\frac{n}{p_i}} \mu(d) 
\cdot (\log d + \log p_i)^m}\end{align}$


The sum of the two logarithms raised to the $m$ power: Replace with the binomial theorem sum.


$\begin{align}{\sum_{d|n} \mu(d) \log^{m+1}d = -\sum_{p_i|n} \log p_i \sum_{d|\frac{n}{p_i}} \mu(d) 
\cdot \sum_{j=0}^m \binom{m}{j} \cdot \log^j d \cdot log^{m-j} p_i}\end{align}$


Home stretch: Swap the order of the inner two sums by first pushing the Möbius function into the innermost loop.


$
\begin{align}
{\sum_{d|n} \mu(d) \log^{m+1}d = -\sum_{p_i|n} \log p_i \sum_{d|\frac{n}{p_i}} 
\sum_{j=0}^m \mu(d) \binom{m}{j} \cdot \log^j d \cdot log^{m-j} p_i}
\end{align}
$


$
\begin{align}
{\sum_{d|n} \mu(d) \log^{m+1}d = -\sum_{p_i|n} \log p_i 
\sum_{j=0}^m 
\sum_{d|\frac{n}{p_i}}
\mu(d) \binom{m}{j} \cdot \log^j d \cdot log^{m-j} p_i}
\end{align}
$


Finally factor elements out of the inner loop that have no $d$ dependence:


$
\begin{align}
{\sum_{d|n} \mu(d) \log^{m+1}d = -\sum_{p_i|n} \log p_i 
\sum_{j=0}^m
\binom{m}{j} \cdot
log^{m-j} p_i \cdot
\sum_{d|\frac{n}{p_i}}
\mu(d) \cdot \log^j d }
\end{align}
$


Here we have an inner loop over divisors with exponent $j$ in the range $0$ to $m$; 
and this we presume is established as zero. (The $m=0$ case is zero by inspection.)


So per the plan at the outset: We have a (roughly speaking) completed proof by induction. 
The argument thread has rested heavily on the separability of the Möbius and logarithm
functions as well as summation logic.  $\;\;$ &#x2610;

#### **2.9 b)** Show $\sum_{d|n} \varphi(\frac{x}{d}, \frac{n}{d}) = \lfloor x \rfloor$.

Partition this into sets corresponding to values of $d$ that divide $n$. Chip uses $A_{xn}(d)$

\textrm{Define } A_d=\{ \; k \in S_x \ni d = (n, k)\; \} \textrm{ and define } f_d = |A_d|
\end{align}$

Of course $k$ will be taking on values that are multiples of $d$ to satisfy that second condition; so 'hopping'. 


As $\{ A_d \}$ partitions $S_x$, the sum of all the $f_d$ values will be $\lfloor x \rfloor$. So far so good. 

Now we can scale down the expression for $d$ in the definition of $f_d$ (really in $A_d$) by
dividing out $d$: The gcd of $n/d$ and $k/d$ becomes $1$ (so they are relatively prime):


$\begin{align}
f_d = \left| \{  \; q \;\; \ni \;\; 1 \le q \le \frac{x}{d} \;\; \cap \;\; 1 = (q, \frac{n}{d}) \; \} \right|
\end{align}$


So $f_d$ is now expressing a count of relative primes precisely in the manner of the extended totient:


$\begin{align}f_d = \varphi \left( \frac{x}{d}, \frac{n}{d} \right) \end{align}$.


Since the $f_d$ summed over divisors of $n$ is $x$ we have 
established that $\sum_{d|n} \varphi \left( \frac{x}{d}, \frac{n}{d} \right) = \lfloor x \rfloor$.  $\;\;$ &#x2610;

Define $S_x = 1, 2, \dots , x $ and partition this based on the divisors $d$ of $n$ into subsets.
For a fixed value of $d \textrm{ we have } A_{n,x,d} = \{ k \backepsilon d = (n, k)\}$.




We have three starter definitions: A set of integers up to 
$\lfloor x \rfloor$, a partition of this set into subsets, and a function based on the partition.


Define $S_x = \{ 1, 2, \dots, \lfloor x \rfloor \}$. Example: $\{ 1, 2, 3, \dots, 17 \}$


Define a partition of $S_x$ into a set of subsets: 
$A_{x, n}(d) = \{ k \in S_x \; { \large \backepsilon } \; (k, n) = d \}$.


English: The $d$-subset of $S_x$ are the numbers that have $gcd$ with respect to $n$ equal to $d$.


Define a function $f$ as the size of $A$: $f_{x, n}(d) = | A_{x, n}(d) |$.


As above with $x = 17.2$ (using $\lfloor x \rfloor = 17$) and $n = 12$


- $A_{17, 12}(1) = \{ 1, 5, 7, 11, 13, 17 \}$; $f = 6$
- $A_{17, 12}(2) = \{ 2, 10, 14 \}$; $f = 3$
- $A_{17, 12}(3) = \{ 3, 9, 15 \}$; $f = 3$
- $A_{17, 12}(4) = \{ 4, 8, 16 \}$; $f = 3$
- $A_{17, 12}(6) = \{ 6 \}$; $f = 1$
- $A_{17, 12}(12) = \{ 12 \}$; $f = 1$



For $n=12$ only the $d \in \{ 1, 2, 3, 4, 6, 12 \}$ yield non-zero $f$ values. 
$A$ partitions $S_x$ without repetitions or skips so the sum of the $f$ values is 
$\lfloor x \rfloor$.

> Note: In 2.10, 2.11, 2.12: $d(n)$ denotes the number of positive divisors of $n$.

### 2.10 Show $\begin{align}{\small\prod_{t|n} t = n^{\frac{d(n)}{2}}}\end{align}$


- $n=1$ checks out...
- Suppose $n$ is prime so $d(n)=2$
    - The RHS becomes ${\large n^{2/2}}$. The LHS product is $1 \cdot n$: checks out...



Let us write the values of $t$ in pairs whose product is $n$. For example for $n=60$ 
there are $12$ factors, $d(60)=12$. Write the left-side product as 
$(1 \cdot 60) \cdot (2 \cdot 30) \cdot \dots \cdot (6 \cdot 10)$. There are $6$ pairs here
so the resulting product is $60^6$. This matches the right side expression $n^{d/2}$; 
so non-perfect-square composites check out...


The last case is for $n$ a perfect square. Let us look at two examples: $n=25$ 
andd $n=36$. We have $d(25)/2 = 3/2$ and $d(36)/2 = 9/2$. The respective left-side products 
are $(1 \cdot 25) \cdot 5$ 
and $(1 \cdot 36) \cdot (2 \cdot 18) \cdot (3 \cdot 12) \cdot (4 \cdot 8) \cdot (6)$.
These are respectively $25^{3/2}$ and $36^{9/2}$; so the case for squares is
"a sequence of product pairs giving $n$ with a final (integer-valued) square root
thrown in". So perfect squares check out and that covers all the cases. &#x2610;


The above solution feels a bit *ad hoc* so I finish by quoting Chip's elegant solution using the index flip: 


$
\begin{align}
n^d = \prod_{t|n}n = \prod_{t|n} t \cdot \frac{n}{t} = \prod_{t|n}{t} \cdot \prod_{t|n}{\frac{n}{t}} = \Bigl\{ \prod_{t|n} t \Bigr\}^2
\end{align}
$
