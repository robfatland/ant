### 2.32 Generalized divisor function $\sigma_\alpha$ 


Prove that 


$\begin{align}
\sigma_\alpha(m) \cdot \sigma_\alpha(n) = \sum_{d|(m,n)} d^\alpha \cdot \sigma_\alpha 
\bigl( \frac{mn}{d^2} \bigr)
\end{align}$

- Note that $\sigma_\alpha$ is multiplicative
- $d$ divides $(m, n)$ means $m = q \cdot d$ and $n = r \cdot d$ and so $mn/d^2$ is a positive integer
- Writing $d^\alpha$ is equivalent to writing $N^{\alpha}(d)$
- We have $((m/d), (n/d)) = (m,n)/d \;\;$ so for the largest $d = (m, n)$ we have

$\begin{align}
d^\alpha \cdot \sigma_\alpha \bigl( \frac{mn}{d^2} \bigr) = 
d^\alpha \cdot \sigma_\alpha \bigl( \frac{m}{d} \bigr) \cdot \sigma_\alpha \bigl( \frac{n}{d} \bigr)
\end{align}$ 

but that is just one term of the sum. 


Chip advises us to solve this by fitting it
into the mold (and result) of problem 2.31. There we showed that 
for $f$ multiplicative and $g$ completely multiplicative, and where
the constituency equation (**CE**)


$\begin{align}
f(p^{n+1}) = f(p)f(p^n) - g(p)f(p^{n-1}) \textrm{ for } n \ge 1
\end{align}$


holds we arrive at this composition:
$f(m) \cdot f(n) = \sum_{d|(m,n)} g(d) \cdot f \left( \frac{mn}{d^2} \right)$.


Need a statement here ... about how the **CE** involves a prime whereas
$m$ and $n$ are arbitrary; so revisit how 2.31 made that transition to invoke
it properly: Once it works for $p$ it can be extended to composites.


So the strategy is to show that the **CE** is correct for all
values of $\alpha$: Sufficient grounds to invoke the 
2.31 result, applied in problem 2.32.


To spell out the correspondences: 
$f \rightarrow \sigma_{\alpha}$ and $g \rightarrow N^{\alpha}$.
First we make sure we are ok with $\alpha = 0$ and then again for $\alpha \ge 1$.


So for $\alpha = 0$ I will evaluate the right side of the **CE**:

$\begin{align}
\sigma_0(p) \cdot \sigma_0(p^n) - N^0(p) \cdot \sigma_0(p^{n-1})
= 2 \cdot (n + 1) - n = n + 2 = \sigma_0(p^{n + 1})
\end{align}$


So the **CE** holds for $\alpha = 0$. Now for $\alpha > 0$: This works out once
we establish the relationship


$\begin{align}
\sigma_\alpha(p^n) = \frac{p^{(n+1)\alpha}-1}{p^{\alpha}-1} \;\; \textrm{ and so } \;\;
\sigma_\alpha(p) = \frac{p^{2\alpha}-1}{p^\alpha-1}
\end{align}$


Let's write out the definition of $\sigma_{\alpha}(p^n)$ to see that this is the case.

$p^n$ has factors $1, p, p^2, \dots, p^n$ where in the generalized divisor function
$\sigma_{\alpha}$ each factor is raised to the $\alpha$ and these are summed. 


$\begin{align}
\sigma_{\alpha}(p^n) = 1 + p^\alpha + p^{2 \alpha} + \dots + p^{n \alpha}
\end{align}$


Ok then as the relationship above suggests: Let us multiply this by $p^\alpha - 1$
to engage a telescope effect. (See problem 2.40 for a reprise of this method.)


$\begin{align}
(p^\alpha - 1) \cdot \sigma_{\alpha}(p^n) 
= p^\alpha + p^{2 \alpha} + \dots + p^{(n+1) \alpha} - 1 - p^\alpha - \dots - p^{n \alpha}
= p^{(n+1) \alpha} - 1
\end{align}$


so by dividing by $p^\alpha - 1$ we get the relationship needed, as noted above.



So here is the **CE** for $\alpha > 0$:


$\begin{align}
\sigma_{\alpha}(p) \cdot \sigma_{\alpha}(p^n) - N^{\alpha}(p) \cdot \sigma_{\alpha}(p^{n-1}) = 
\frac{(p^{2 \alpha}-1)}{p^{\alpha}-1} \cdot 
\frac{(p^{(n+1)\alpha}-1)}{p^{\alpha}-1} - p^{\alpha} \cdot \frac{(p^{n \alpha}-1)}{p^{\alpha}-1}
\end{align}$


The first term in the right side expression has a 'difference of two squares' numerator so
this can be factored with one factor canceling the respective denominator; so the right side
can be consolidated as


$\begin{align}
\sigma_{\alpha}(p) \cdot \sigma_{\alpha}(p^n) - N^{\alpha}(p) \cdot \sigma_{\alpha}(p^{n-1}) = 
\frac
{(p^{\alpha}+1) \cdot (p^{(n+1)\alpha}-1) - p^{\alpha} \cdot (p^{n \alpha} -1)}
{p^{\alpha}-1}
= \frac{p^{(n + 2) \alpha} - p^\alpha + p^{(n+1)\alpha} - 1 - p^{(n + 1)\alpha} + p^\alpha}{p^{\alpha}-1}
= \frac{p^{(n + 2) \alpha} - 1}{p^\alpha - 1}
= \sigma_{\alpha}(p^{n+1}).
\end{align}$


This establishes the **CE** for all $\alpha \ge 0$; and according to the rationale given above: The
**CE** holding 'fits the model' of problem 2.31, so completing the proof of 2.32.  &#x2610;  

### 2.33 Liouville function $\lambda(n)$


This arithmetic function $\lambda(n)$ has value $-1$ or $+1$ giving the prime 
factor parity of $n$ including multiplicity. That is: Write $n$ as the
product of primes $p_i$ raised to powers $a_i$; then we have a count of primes 
$\Omega(n)$ as the sum of all the $a_i$; and then $\lambda = (-1)^\Omega$.


This problem is to prove Liouville's function can be written


$\begin{align}
\lambda(n) = \sum_{q^2 | n} \mu \bigl( {\frac{n}{q^2}} \bigr)
\end{align}$



This intended form for $\lambda$ looks like a Dirichlet convolution with a 
constrained set of divisors; and the relevant theorem invokes
the "square indicator function" (one can also say "square characteristic function") 
which is $1$ for $n$ a perfect square and $0$ otherwise; so calling this
function $f(n)$ let's begin with this theorem.


Theorem 2.19 gives us $f(n)$ as a sum over $d|n$ of $\lambda(d)$ which in
Dirichlet convolution notation is $f = \lambda * u$. This uses $u(n) = 1$ 
so this first step is a finesse from a sum over divisors of $n$ of $\lambda$ 
to the notation of Dirichlet convolution. Next to employ $\mu$
as the inverse of $u$:


$\begin{align}
\lambda * u = f
\end{align}$


$\begin{align}
\lambda * u * \mu = \lambda = f * \mu
\end{align}$


$\begin{align}
\lambda = \sum_{d|n} f(d) \cdot \mu \left( {\frac{n}{d}} \right)
\end{align}$


The terms of this sum will be zero unless the argument of $f(d)$ is a square.
So let's constrain divisors $d$ to those that can be written as a square $q^2$. 


$\begin{align}
\lambda = \sum_{{q^2}|n} f({q^2}) \cdot \mu \bigl( {\frac{n}{q^2}} \bigr) =
\sum_{{q^2}|n} \mu \bigl( {\frac{n}{q^2}} \bigr)
\end{align}$

### 2.35 Prove $h(n)$ as defined below is multiplicative


Defining $h(n)$: Let $f$ and $g$ be multiplicative functions and $a$ and $b$ 
positive integers with $a \ge b$.


$\begin{align}
\textrm{Define } h(n) = \sum_{d^a|n} f \Bigl( \frac{n}{d^a} \Bigr) \; g \Bigl( \frac{n}{d^b} \Bigr)
\end{align}$


Here $d$ is a divisor of $n$ but the sum is constrained to only $a$th powers of $d$ that divide $n$.
As $b$ is less than or equal to $a$: In such cases $n$ is divisible by $d^b$ as well.
Take $n=24$: Then for $a=2$ the sum would be over $\{1, 4\}$ and for $a=3$ the sum would be over $\{1, 8\}$.

The plan is obvious enough: Show that for $(m, n) = 1 \textrm{ we have } h(m) \cdot h(n) = h(m \cdot n)$.
This seems to be the downhil direction in Chip's solution. I also follow his notational point of departure:
From $(m, n) = 1$ we can imagine divisors $c$ and $d$ of $m$ and $n$ respectively for which it will also be
the case that $(m/c, n/d) = 1$. 


$\begin{align}
h(m) \cdot h(n) = 
\sum_{c^a|m} f \bigl( \frac{m}{c^a} \bigr) g \bigl( \frac{m}{c^b} \bigr) \cdot 
\sum_{d^a|n} f\bigl( \frac{n}{d^a} \bigr) g \bigl( \frac{n}{d^b} \bigr)
\end{align}$


Multiplying these two sums produces a grand sum over all allowed pairs of $c^a$ and $d^a$; where with $(c, d) = 1$
this is equivalent to indexing all possible composite pairs $e = c \cdot d$.


$\begin{align}
h(m) \cdot h(n) = \; &
\sum_{c^a|m \atop d^a|n} f \bigl( \frac{m}{c^a} \bigr) \cdot g \bigl( \frac{m}{c^b} \bigr) \cdot 
f\bigl( \frac{n}{d^a} \bigr) \cdot g \bigl( \frac{n}{d^b} \bigr) \\
= \; & \sum_{c^a|m \atop d^a|n} f \bigl( \frac{m \cdot n}{(cd)^a} \bigr) \cdot g \bigl( \frac{m \cdot n}{(cd)^b} \bigr) \\
= \; & \sum_{e^a|mn} f \bigl( \frac{m \cdot n}{e^a} \bigr) \cdot g \bigl( \frac{m \cdot n}{e^b} \bigr) \\
= \; & h(m \cdot n)
\end{align}$

### Introducing Möbius functions of order $k \ge 1$


Define $\mu_k(n)$ by cases dependent on the factorization of $n$:


- Case 1: $\mu_k(1) = 1$
- Case 2: $\mu_k(n) = 0$ when any $p^{k+1}|n$
- Case 3: $\mu_k(n) = (-1)^r$ if $n= \prod_{j=1}^r p_j^k \cdot \prod_{i} p_i^{a_i < k}$
- Case 4: $\mu_k(n) = 1$ otherwise


Analogs to $\mu(n)$: We have $\mu_k(1)=\mu(1)=1$. 
$\mu(n)=0$ when $n$ has a square divisor; and $\mu_k(n)=0$ when
$n$ has a *power-$(k+1)$* divisor.
When $n$ fails to achieve winning factors $p^k: \; \mu_{k}(n)$ behaves like $\mu_k(1) = 1$.

### 2.36 Show that for $k \ge 1$ we have $\mu_k(n^k) = \mu(n)$


With $k=1$ a trivial equivalence, we are concerned with $k > 1$; and the point of 
the problem is that the $n^k$ argument in an order-$k$ Möbius functions has
a *constrained* prime factorization; so $\mu_k(n^k)$ matches
$\mu(n)$. I show this equality condition 'is fine' by cases:



Case 1: $n=1$ is fine as $\mu_k(1^k) = \mu(1) = 1$.


Case 2: $n$ is not square free; $n$ has divisor $p^{q \; \ge \; 2}$ so $\mu(n)=0$. 
This is fine as $p^{(k \cdot q) \; \gt \; k} \; \big| \; n^k$ so $\mu_k(n^k) = 0$.


Case 3: Suppose $n$ is square free: $n = \prod_{i=1}^{r} p_i$. 
Then $n^k = \prod_{i=1}^r p_i^k$ and $\mu(n) = (-1)^r = \mu_k(n^k)$.


There are no other cases available so we are done.$\;\;$ &#x2610; 


Chip's more rigorous solution is an induction proof on $k$.

### 2.37 Show that $\mu_k(n)$ is multiplicative.


I proceed by cases on $\mu_k(m \cdot n)$ where $(m, n) = 1$. 


(Aside, $\mu_k$ is not *completely* multiplicative by counterexample: $\mu_3(27) = -1 \ne \mu_3(9) \cdot \mu_3(3)$.)


Case 0: For $m = 1$ we have $\mu_k(m \cdot n) = 1 \cdot \mu_k(n)$. 


For both $m, n > 0$: 

Case 1: $p^{(a>k)} \; | \; m \implies \mu_k(m \cdot n) = 0.$

Case 2: There is no prime $p$ such that $p^k \; | \; m \; \textrm{ or } \; n$. 
In this case $\mu_k(m \cdot n) = 1 = \mu_k(m) \cdot \mu_k(n)$.

Case 3: $k$-powers of prime factors are present in amounts $q$ and $r$ for $m$ and $n$ respectively.
Then $\mu_k(m \cdot n) = (-1)^{q + r} = \mu_k(m) \cdot \mu_k(n)\;\;\;\;\;\;\;$ &#x2610;

### 2.38  Show that if $k \ge 2$ then 


$\begin{align}
\mu_k(n) = \sum_{d^k|n} \mu_{k-1}\Bigl( \frac{n}{d^k} \Bigr) \mu_{k-1}\Bigl( \frac{n}{d} \Bigr)
\end{align}$


Both sides are multiplicative (see problems 2.37 and 2.35 for this result, lhs and rhs respectively).
The rhs could probably use a little more checking. But to continue: With multiplicative functions it
will suffice to show the desired result for argument $p^a$. 


The case of $a=0$ so $n=1$ checks out: Both sides are $1$.


For the case where $n = p^{k+1}$: The lhs evaluates to $0$. 
In the rhs sum: We have two terms to consider, $d=1$ and $d=p$ with the 
righthand $\mu_{k-1}$ having corresponding arguments $p^{k+1}$ and 
$p^k$. These both evaluate as $0$, case closed.


This leaves two remaining cases: $n = p^k$ and $n = p^{j < k}$. For both: The lhs $= 1$.

For $n = p^k$ the rhs is two nonzero terms: $d=1$ and $d=p$. 
This gives $1 \cdot 0 + (-1) \cdot \mu_{k-1}(p^{k-1}) = -1 \cdot -1 = 1$, correct. 

For $n = p^j$ the rhs is one terms, $d = 1$. This evaluates as $1 \cdot 1 = 1$, correct. &#x2610;

### 2.39 Show that if $k \ge 1$ then $\bigl| \mu_k(n) \bigr| = \sum_{d^{k+1}|n}\mu(d)$

Problem 2.6 proves (slightly paraphrased):


$\begin{align}
\sum_{d^{k+1}|n} \mu(d) = \begin{cases}
0 & \textrm{ if } m^{k+1}|n \textrm{ for some } m > 1 \\
1 & \textrm{ otherwise }
\end{cases}
\end{align}$

$\begin{align}
\mu(n) = \begin{cases} 
1 & \textrm{ if } n = 1 \\
-1^k & \textrm{ if } n \textrm{ is the product of } k \textrm{ distinct primes } \\
0 & \textrm{ otherwise }
\end{cases}
\end{align}$


Let's check the four cases for argument $n$:


- Case 1 $\;\;n = 1$ so $|\mu_{k}(1)| = 1$, confirmed, done.
- Case 2 $\;\;n$ has a factor $p^{k+1}$ so $|\mu_{k}(n)| = 0$, RHS is likewise $0$ from the problem 2.6 result, done.
- Case 3 $\;\;n$ has $r$ factors of the form $p^k$ but no higher exponents: $|\mu_{k}(n)| = 1$, RHS has only $d=1$, confirmed, done.
- Case 4 $\;\;n$ has only factors $p^{j < k}$: $|\mu_{k}(n)| = 1$, RHS has only $d=1$, confirmed, done.

### 2.40 Show that for each prime $p$ the Bell series for $u_k$ is given by 


$\begin{align}(\mu_k)_p(x) = \frac{1-2x^k+x^{k+1}}{1-x}\end{align}$


This is interesting as the choice of prime $p$ evaporates from the end result.


The Bell series of $\mu_k \textrm{ mod } p$ is defined in this case


$\begin{align}(\mu_k)_p(x) = \sum_{n=0}^{\infty} \mu_k(p^n)x^n\end{align}$


What follows is valid for $k=1$ but is written more in the spirit of $k>1$. 
The argument of the infinite sum becomes zero when $n$ exceeds $k$
so the sum can be truncated at $k$; and then rewritten in a more concrete form: 


$\begin{align}(\mu_k)_p(x) = \sum_{n=0}^{k} \mu_k(p^n)x^n = 1 + x + \sum_{n=2}^{k} \mu_k(p^n)x^n\end{align}$


The prime $p$ is raised to a value less than $k$ for most of this sum, i.e. $\mu_k = 1$;
and the final term $n=k$ resolves to $(-1)^1$:


$\begin{align}(\mu_k)_p(x) = \sum_{n=0}^{k} \mu_k(p^n)x^n = 1 + x + \sum_{n=2}^{k-1} x^n - x^k\end{align}$


Observing the pattern:


$\begin{align}k=1: (\mu_k)_p(x) = 1-x\end{align}$


$\begin{align}k=2: (\mu_k)_p(x) = 1 + x - x^2\end{align}$


$\begin{align}k=3: (\mu_k)_p(x) = 1 + x + x^2 - x^3\end{align}$


$\begin{align}k=4: (\mu_k)_p(x) = 1 + x + x^2 + x^3 - x^4\end{align}$


$\begin{align}k:(\mu_k)_p(x) = 1 + x + x^2 + \dots + x^{k-1} - x^k\end{align}$


In the spirit of a telescoping series: Multiply the last (general) expression above by $1-x$: 


$\begin{align}(1-x) \cdot (1 + x + x^2 + \dots + x^{k-1} - x^k) =  
1 + x + x^2 + \dots + x^{k-1} - x^k - x - x^2 - x^3 - \dots - x^{k-1} - x^{k} + x^{k+1} = 1 - 2x^k + x^{k+1}\end{align}$


So the general expression can be written 


$\begin{align}1 + x + x^2 + \dots + x^{k-1} - x^k = \frac{1 - 2x^k + x^{k+1}}{1-x} = (\mu_k)_p(x).\;\;\end{align}$ &#x2610;
