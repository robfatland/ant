## Introduction to Analytic Number Theory Chapter 1


### Useful ideas:


* $a|b$ meaning 'a divides b' is equivalent to $\; \exists \;\; q \; \backepsilon b = a \cdot q$: There exists an 
integer $q$ for which $b$ is the product $aq$.


* $(a, b)$ the greatest common divisor (gcd) of $a$ and $b$ is a positive integer. 


* $(a, b) = 1$ means $a$ and $b$ are relatively prime. 


* $(a, b) = c \implies \; \exists \; x \textrm{ and } y \backepsilon c = xa + yb$.
$x$ and $y$ exist (and in passing: $(x, y) = 1$). These values $x$ and $y$ can be 
determined for example by back-propagating the Euclidean
algorithm for finding the gcd (see problem 1.20). Two important qualifiers concerning this linear combination of $a$ and $b$:
    * For an *arbitrary* $x$ and $y$: $x \cdot a + y \cdot b$ can't be a positive 
common factor of $a$ and $b$ less than $(a, b)$. 
    * $xa + yb$ will be a *multiple* of $(a, b)$.<BR><BR>
    

Notice there is some bi-directionality here: The gcd is the *greatest* common
divisor but the smallest possible positive $x \cdot a + y \cdot b$ will be this gcd. 


Example: Suppose $a=12$ and $b=30$. We see that 
the gcd is $c = 6$ where $a = 2 \cdot c$ and $b = 5 \cdot c$. A
linear combination from some $x$ and some $y$ will be a construction
in 'units' of the gcd, a basic building block.
There is no linear combination that arrives at a smaller positive value; no $xa + yb$ 
will be found to equal $3$ or $2$ or $1$. This shared building block concept
is a very central idea: $c=(a,b)$ replaces $1$ as a new atomic unit ($c$)
that re-scales the number line. 
    
    
Note for puzzle solvers: This atomic nature of the gcd is apparent in 
un-solvable gallon puzzles such as 'You have two containers with capacities 
four gallons and six gallons. The horses need precisely five gallons of water
in their watering trough.' Let's just give the horses six gallons of water 
and endure their wry looks.


* $x$ and $y$ always exist such that the expression $ax + by$ can be substituted 
for $(a, b)$.
* $x$ and $y$ are not unique: We can add/subtract multiples of zero.
    * Example: $a=15, \; b=24$. Then $c = (a,b) = 3 = 15\cdot(-3) + 24\cdot (2)$.
Since any $n$ gives $n \cdot (8a-5b)=0$ we have values for $x: \; -3+8 \cdot n$ and 
$y: \; 2 - 5 \cdot n$ that also yield $a \cdot x + b \cdot y = 3$.


For more on this gcd thought process see the remark in problem 1.2.
    

* If $a | b$ and $a | (b + c)$ then $a | c$; see problem 1.18.

#### Intermezzo on the gcd


These remarks interrupt the problem solutions to reinforce the greatest common divisor 
core concepts; so intentionally repetitive, small bites.
    
    
We are concerned with a divisor of two numbers.


3 is a divisor of 12 and 30. 
    
    
1 divides everything. 


A number $a$ has a set of divisors. So does $b$. The intersection of these two 
sets is the set of common divisors of $a$ and $b$.

    
There is always a largest element in the set of common divisors of two numbers. 
For 12 and 30 that is 6. When it is $1$ (say for $3$ and $7$) the two numbers
are *relatively prime*.

Theorem 1.2: A greatest common divisor of $a$ and $b$ exists and can be 
expressed as a linear combination of $a$ and $b$: $gcd = ax + by$.


This is a building block idea: Both $a$ and $b$ are constructed of
common divisor 'blocks' or 'atoms' in some quantity. 

    
When comparing $a$ and $b$ we see this gcd
building block view. We can always take integer multiples of both $a$ and $b$ (i.e. $x$ and $y$
amounts), to arrive at just a single gcd block. This is an interpretation of
$gcd = ax + by$. We can't, however, use other values of $x$ and $y$ to arrive at a number smaller 
than the gcd.


A mistake: 'Any arbitrary linear combination of $a$ and $b$
is a divisor of $a$ and $b$.' Take $(3, 7) = 1$ for example: $3\cdot3 + (-1)\cdot7=2$,
not a common divisor of $3$ and $7$. $2$ is however a *multiple* 
of the gcd of $3$ and $7$ (which is $1$). 


A correction of the mistake: 'Any arbitrary linear combination of $a$ and $b$
is a multiple of the gcd of $a$ and $b$.' 


Getting a linear combination of $a$ and $b$ to equal an inferior common divisor
can't be done because $a$ and $b$ are always separated by some multiple of the gcd.


Result: The gcd of $a$ and $b$ can be expressed as a linear combination of $a$
and $b$; or some *multiple* of the gcd can be expressed this way; but nothing
else. Let's take $a=18$ and $b=30$ with a gcd of 6. The linear combination 
$4\cdot a-2 \cdot b=72-60=12=2\cdot6$ is twice the gcd.  

A more general take on producing *multiples* of the
gcd: Multiply the gcd-specific "x and y" 
linear combination by any factor. 


$a=2 \cdot 3 \cdot 11$ and $b=2 \cdot 5 \cdot 7$.


$(66, 70) = 2 = 66 \cdot x + 70 \cdot y$. Values for $x$ and $y$ that
produce the gcd are $x=-18$ and $y=17$.


Multiply the gcd expression by $3$ to arrive at a new linear combination of 
$a$ and $b$ equal to $6$, three times the gcd. This $6$ is not a common divisor 
of $66$ and $70$ but it *is* a linear combination that is less than both of them.


Additional observation: 
Suppose $(a, b) = d$; then the $x$ and $y$ values in
$ax + by = d$ are not unique. They can be modified for example using a 
free parameter $n$. First we have $ax + by + nab - nab = d$.
Reorganized this becomes $a(x + nb) + b(y - na) = d$. 


There is consequently an infinite set of workable $\{x, y\}$ pairs to produce
the gcd. This is used in problem 1.25.


Any linear combination of $a$ and $b$ will be a multiple of $(a, b)$. 
Therefore $(a,b)|(ax+by)$ for any $x$ and any $y$.


We can ask: How to solve for $(a, b)$? Suppose I have a linear combination 
of $a$ and $b$ equal to some number $q$ less than both $a$ and $b$. Then I come up with 
another number $r$ less than $q$ from a different linear combination. This lowers
the upper bound on the gcd: Now it is $r$. If $q$ is not a multiple of $r$ then they 
have some common divisor that is smaller still that would be the new upper bound. If this
gcd of $q$ and $r$ happens to be $1$ then $a$ and $b$ are relatively prime. 


This suggests inventing a problem: Use
two (or more) predetermined $\{ x, y \}$ pairs to create linear combinations of two numbers
$a$ and $b$; and from that information derive $(a, b)$.

#### 1.3 If $(a, b) = 1$ then $(a^n,b^k)=1$ for all $n \ge 1, \; k \ge 1$.

Both $a$ and $b$ have prime factorizations with no primes in common. $a^n$ can be seen as 
the product of the '$a$' primes each raised to the $n$ power, and likewise for $b^k$. 
Hence the factorizations of $a^n$ and $b^k$ have no primes in common and the resulting gcd is $1.\;\;$ &#x2610;

#### 1.4 If $(a, b) = 1$ then $(a+b, a-b)$ is either $1$ or $2$.

For $a=10 \textrm{ and } b=7 \textrm{ we have }(17,3)=1$. For
$a=9 \textrm{ and } b=7 \textrm{ we have }(16, 2) = 2$. So
$(a+b, a-b)$ can be $1$ or $2$. It remains to show $c = (a+b, a-b)$ 
can not exceed $2$.

Expand $c = (a+b,a-b) > 2$ as $a + b = \alpha c \textrm{ and } a - b = \beta c$.
Add these two terms together: $a + b + a - b = 2a = c(\alpha + \beta) \textrm { and therefore } c | 2a$.
Since this presumed $c$ is greater than $2$ it can not divide $2$ so it must divide $a$: $c | a$. Since
$c | a$ and $c | a + b$: $c$ will also divide $b$; but this is not possible as $(a, b) = 1$. 
Consequently $(a + b, a - b)$ can only take on values $1$ and $2.\;\;$ &#x2610;

#### 1.5 Show that if $(a,b)=1$ then $(a+b, a^2-ab+b^2)$ is either $1$ or $3$.

Use $c=(a+b, a^2-ab+b^2)$.
Take $a=4 \textrm{ and }b=3: \; (7,13)=1$.
Take $a=5 \textrm{ and } b=7: \; (12, 39)=3$. 
Both $c=1$ and $c=3$ are possible gcd values.


Expand $c = (a+b, a^2-ab+b^2) \textrm{ as two expressions: }a+b=c \alpha \textrm{ and }a^2-ab+b^2 = c \beta$.
Square the first and subtract the second: $3ab=c(c \alpha^2 - \beta) = c \gamma$. Note $\gamma \ne 0$.


Now taking $c > 1$ write $c = p \cdot q$ where $p$ is a prime and $q \ge 1$. Since $3 \cdot a \cdot b = p \cdot q \cdot \gamma$
it must be the case that $p$ divides one of $\{ 3, a, b \}$.


Since $c | (a + b)$ it is also the case that $p | (a + b)$. Suppose $p | a$: Then since $p | (a + b)$, we can conclude that $p | b$ and this
is a contradiction of $(a, b) = 1$. Likewise we reach the same contradiction by supposing $p | b$. So $p$ can only take
on the value $3$. In this case $q | a \cdot b$ which runs into the same contradiction unless $q=1$. This gives $c = 3$ as the
only workable value for $c > 1$. Consequently $c$ is either $1$ or $3.\;\;$ &#x2610;

#### 1.6 Show that if $(a, b)=1$ and $d|(a+b)$ then $(d,a)=(d,b)=1$.

Set $(d,a)=c$. Then $d = c \alpha \textrm{ and } a = c \beta$.
As $d|(a + b)$ it is also the case that $c | (a + b): \; c \gamma = a + b = c \beta + b.$
Hence $b = c(\gamma - \beta) \textrm { meaning that } c|b$. So $c$, the gcd of $a$ and $d$, 
divides both $a$ and $b$. 
This means $(a, b) \ge c$ but we are given that $(a, b) = 1$ so $c = 1 = (d, a)$.


As there is nothing special about starting with $(d, a)$ we can use the
same argument to arrive at $(d, b) = 1.\;\;$ &#x2610;

#### 1.7 If the sum of two reduced fractions is an integer, say $a/b + c/d = n$ where $(a,b)=(c,d)=1$, show that $|b|=|d|$. 


$\begin{align}n = \frac{ad + bc}{bd} \implies bdn = ad + bc \implies \textrm{ both } b|(ad+bc) \textrm{ and } d|(ad+bc)\end{align}$.




Now consider that $b$ divides a sum *and* it divides the second term $bc$.
Therefore it must divide the first term as well, $ad$. So $b|ad$ but $(a, b) = 1$
so $b|d$. By the same reasoning $d|b$ so with this reciprocal divisibility we conclude $|b|=|d|.\;\;$ &#x2610;


As noted this facilitates problem 1.13.

#### 1.8 A number is *square-free* if it is not divisible by the square of any prime. Prove that for every $n \ge 1$ there exist unique $a > 0$ and $b > 0$ such that $n = a^2 b$ with b *square-free*.

The numbers $a$ and $b$ can be built by a decomposition of each
$\alpha_{i}$ prime exponent of $n$ as 
$2 \cdot \lfloor {\alpha_i/2} \rfloor + \alpha_{i} \; mod \; 2$.
We then have $a=\prod p_i^{\lfloor {\alpha_i/2} \rfloor} \textrm{ and }
b=\prod p_i^{\alpha_{i} \; mod \; 2}$ so that $n=a^2b.\;\;$ &#x2610;


To demonstrate this factorization is unique frankly I'd just 
cite the FTA. I suppose we could also say that any migration
of a prime factors from $b$ to $a^2$ is impossible since $b$
is squarefree; and likewise in the other direction.

#### 1.9 b) Likewise: If $b^2$ is the largest square divisor of $n$, then $a^2 | n$ implies $a | b$. 

True by decomposition: As in **1.8** we can manufacture $b$ from the prime factorization of $n$:

#### 1.10 Given $x$ and $y$, let $m = ax + by$ and $n = cx + dy$ where $ad - bc = \pm 1$. Prove $(m, n) = (x, y)$.


Part 1: Looking ahead we need a general
result, taking $q, r, s > 0$ to avoid absolute value clutter:
If $q|r$ and $q|s$ then $q |(r, s)$. This is easy to 
show: $q|r \implies r = q \alpha$ and likewise $s = q \beta$.
Then $(r, s) = (q \alpha, q \beta) = q (\alpha, \beta)$ from theorem 1.3c; 
so $q|(r,s)$.


Part 2: 
The condition imposed on 
$a, b, c, \textrm{and} \; d$ 
suggests the determinant of a 2D linear transform. 
From linear algebra: The inverse of a $2 \times 2$ matrix
$
A=\left[{\begin{array}{cc}
   a & b \\
   c & d \\
  \end{array}}\right]
\; \; \; $
is
$\; \; \; A^{-1} = \frac{1}{det A} \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]$ with $det A = ad - bc$.
  
  
Writing $x$ and $y$ as a column vector
$\tilde{v}=\left[{\begin{array}{c}x\\y\end{array}}\right]$
and the result of the transform as
$\tilde{k}=\left[{\begin{array}{c}m\\n\end{array}}\right]$
we have $A \tilde{v} = \tilde{k}$ and $\tilde{v}=A^{-1}\tilde{k}$.

$\left[{\begin{array}{c}x\\y\end{array}}\right] =
\frac{1}{det A} \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]
\left[{\begin{array}{c}m\\n\end{array}}\right] =
\pm 1 \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]
\left[{\begin{array}{c}m\\n\end{array}}\right] 
$

In rude mechanical terms
$x = \pm (md-nb) \textrm{ and } y = \pm (na-mc)$.
However we do not need to write this out as such.
Rather say: 
$x$ and $y$ are linear combinations of $m$ and $n$
so $(m, n)|x$ and $(m, n)|y$. This in turn (from part 1)
means that $(m, n)|(x, y)$. 

In the problem statement we see $m$ and $n$ are linear combinations 
of $x$ and $y$. By the same argument then $(x, y)|(m, n)$.
As $(x, y)|(m, n)$ and $(m, n)|(x,y)$ the two are equal,
$(m, n)=(x,y).\;\;$ &#x2610;
