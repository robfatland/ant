## On the gcd


These remarks reinforce the greatest common divisor 
core concepts; so intentionally repetitive, small bites.
    
    
We are concerned with a divisor of two numbers.


3 is a divisor of 12 and 30. 
    
    
1 divides everything. 


A number $a$ has a set of divisors. So does $b$. The intersection of these two 
sets is the set of common divisors of $a$ and $b$.

    
There is always a largest element in the set of common divisors of two numbers. 
For 12 and 30 that is 6. When it is $1$ (say for $3$ and $7$) the two numbers
are *relatively prime*.

Theorem 1.2: A greatest common divisor of $a$ and $b$ exists and can be 
expressed as a linear combination of $a$ and $b$: $gcd = ax + by$.


This is a building block idea: Both $a$ and $b$ are constructed of
common divisor 'blocks' or 'atoms' in some quantity. 

    
When comparing $a$ and $b$ we see this gcd
building block view. We can always take integer multiples of both $a$ and $b$ (i.e. $x$ and $y$
amounts), to arrive at just a single gcd block. This is an interpretation of
$gcd = ax + by$. We can't, however, use other values of $x$ and $y$ to arrive at a number smaller 
than the gcd.


A mistake: 'Any arbitrary linear combination of $a$ and $b$
is a divisor of $a$ and $b$.' Take $(3, 7) = 1$ for example: $3\cdot3 + (-1)\cdot7=2$,
not a common divisor of $3$ and $7$. $2$ is however a *multiple* 
of the gcd of $3$ and $7$ (which is $1$). 


A correction of the mistake: 'Any arbitrary linear combination of $a$ and $b$
is a multiple of the gcd of $a$ and $b$.' 


Getting a linear combination of $a$ and $b$ to equal an inferior common divisor
can't be done because $a$ and $b$ are always separated by some multiple of the gcd.


Result: The gcd of $a$ and $b$ can be expressed as a linear combination of $a$
and $b$; or some *multiple* of the gcd can be expressed this way; but nothing
else. Let's take $a=18$ and $b=30$ with a gcd of 6. The linear combination 
$4\cdot a-2 \cdot b=72-60=12=2\cdot6$ is twice the gcd.  

A more general take on producing *multiples* of the
gcd: Multiply the gcd-specific "x and y" 
linear combination by any factor. 


$a=2 \cdot 3 \cdot 11$ and $b=2 \cdot 5 \cdot 7$.


$(66, 70) = 2 = 66 \cdot x + 70 \cdot y$. Values for $x$ and $y$ that
produce the gcd are $x=-18$ and $y=17$.


Multiply the gcd expression by $3$ to arrive at a new linear combination of 
$a$ and $b$ equal to $6$, three times the gcd. This $6$ is not a common divisor 
of $66$ and $70$ but it *is* a linear combination that is less than both of them.


Additional observation: 
Suppose $(a, b) = d$; then the $x$ and $y$ values in
$ax + by = d$ are not unique. They can be modified for example using a 
free parameter $n$. First we have $ax + by + nab - nab = d$.
Reorganized this becomes $a(x + nb) + b(y - na) = d$. 


There is consequently an infinite set of workable $\{x, y\}$ pairs to produce
the gcd. This is used in problem 1.25.


Any linear combination of $a$ and $b$ will be a multiple of $(a, b)$. 
Therefore $(a,b)|(ax+by)$ for any $x$ and any $y$.


We can ask: How to solve for $(a, b)$? Suppose I have a linear combination 
of $a$ and $b$ equal to some number $q$ less than both $a$ and $b$. Then I come up with 
another number $r$ less than $q$ from a different linear combination. This lowers
the upper bound on the gcd: Now it is $r$. If $q$ is not a multiple of $r$ then they 
have some common divisor that is smaller still that would be the new upper bound. If this
gcd of $q$ and $r$ happens to be $1$ then $a$ and $b$ are relatively prime. 


This suggests inventing a problem: Use
two (or more) predetermined $\{ x, y \}$ pairs to create linear combinations of two numbers
$a$ and $b$; and from that information derive $(a, b)$.
