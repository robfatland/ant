## Introduction to Analytic Number Theory Chapter 1


### Useful ideas:


* $a|b$ meaning 'a divides b' is equivalent to $\; \exists \;\; q \; \backepsilon b = a \cdot q$: There exists an 
integer $q$ for which $b$ is the product $aq$.


* $(a, b)$ the greatest common divisor (gcd) of $a$ and $b$ is a positive integer. 


* $(a, b) = 1$ means $a$ and $b$ are relatively prime. 


* $(a, b) = c \implies \; \exists \; x \textrm{ and } y \backepsilon c = xa + yb$.
$x$ and $y$ exist (and in passing: $(x, y) = 1$). These values $x$ and $y$ can be 
determined for example by back-propagating the Euclidean
algorithm for finding the gcd (see problem 1.20). Two important qualifiers concerning this linear combination of $a$ and $b$:
    * For an *arbitrary* $x$ and $y$: $x \cdot a + y \cdot b$ can't be a positive 
common factor of $a$ and $b$ less than $(a, b)$. 
    * $xa + yb$ will be a *multiple* of $(a, b)$.<BR><BR>
    

Notice there is some bi-directionality here: The gcd is the *greatest* common
divisor but the smallest possible positive $x \cdot a + y \cdot b$ will be this gcd. 


Example: Suppose $a=12$ and $b=30$. We see that 
the gcd is $c = 6$ where $a = 2 \cdot c$ and $b = 5 \cdot c$. A
linear combination from some $x$ and some $y$ will be a construction
in 'units' of the gcd, a basic building block.
There is no linear combination that arrives at a smaller positive value; no $xa + yb$ 
will be found to equal $3$ or $2$ or $1$. This shared building block concept
is a very central idea: $c=(a,b)$ replaces $1$ as a new atomic unit ($c$)
that re-scales the number line. 
    
    
Note for puzzle solvers: This atomic nature of the gcd is apparent in 
un-solvable gallon puzzles such as 'You have two containers with capacities 
four gallons and six gallons. The horses need precisely five gallons of water
in their watering trough.' Let's just give the horses six gallons of water 
and endure their wry looks.


* $x$ and $y$ always exist such that the expression $ax + by$ can be substituted 
for $(a, b)$.
* $x$ and $y$ are not unique: We can add/subtract multiples of zero.
    * Example: $a=15, \; b=24$. Then $c = (a,b) = 3 = 15\cdot(-3) + 24\cdot (2)$.
Since any $n$ gives $n \cdot (8a-5b)=0$ we have values for $x: \; -3+8 \cdot n$ and 
$y: \; 2 - 5 \cdot n$ that also yield $a \cdot x + b \cdot y = 3$.


For more on this gcd thought process see the remark in problem 1.2.
    

* If $a | b$ and $a | (b + c)$ then $a | c$; see problem 1.18.

#### 1.3 If $(a, b) = 1$ then $(a^n,b^k)=1$ for all $n \ge 1, \; k \ge 1$.

Both $a$ and $b$ have prime factorizations with no primes in common. $a^n$ can be seen as 
the product of the '$a$' primes each raised to the $n$ power, and likewise for $b^k$. 
Hence the factorizations of $a^n$ and $b^k$ have no primes in common and the resulting gcd is $1.\;\;$ &#x2610;

#### 1.4 If $(a, b) = 1$ then $(a+b, a-b)$ is either $1$ or $2$.

For $a=10 \textrm{ and } b=7 \textrm{ we have }(17,3)=1$. For
$a=9 \textrm{ and } b=7 \textrm{ we have }(16, 2) = 2$. So
$(a+b, a-b)$ can be $1$ or $2$. 


Shorthand: Define $c = (a+b, \; a-b)$. The proof consists in showing $c < 3$.


Lets presume $c > 2$ and introduce two non-zero parameters $\alpha$ and $\beta$: $a + b = \alpha c$ and $a - b = \beta c$.


Add these two terms together: $a + b + a - b = 2a = c(\alpha + \beta)$. This implies $c | 2a$.


Now subtract them: $a + b - (a - b) = 2b = c(\alpha - \beta)$. This implies $c | 2b$.


Suppose that $c > 2$ is odd: Then $c | 2a$ with $(c, 2) = 1$ so we have $c | a$. 
Since $c | a$ and $c | (a + b)$: $c$ also divides $b$; but this is not possible as $(a, b) = 1$.


Suppose that $c > 2$ is even: Write $c = 2 \cdot e$ for some $e > 1$ (because $c > 2$). 
Now we have $2e | 2a$ and $2e | 2b$. Now (taking out the 2s) $e | a$ and $e | b$ which likewise contradicts $(a, b) = 1$. 


Consequently $(a + b, a - b)$ can only take on values $1$ and $2.\;\;$ &#x2610;

#### 1.5 Show that if $(a,b)=1$ then $(a+b, a^2-ab+b^2)$ is either $1$ or $3$.

Use $c=(a+b, a^2-ab+b^2)$.
Take $a=4 \textrm{ and }b=3: \; (7,13)=1$.
Take $a=5 \textrm{ and } b=7: \; (12, 39)=3$. 
Both $c=1$ and $c=3$ are possible gcd values.


Expand $c = (a+b, a^2-ab+b^2) \textrm{ as two expressions: }a+b=c \alpha \textrm{ and }a^2-ab+b^2 = c \beta$.
Square the first and subtract the second: $3ab=c(c \alpha^2 - \beta) = c \gamma$. Note $\gamma \ne 0$.


Now taking $c > 1$ write $c = p \cdot q$ where $p$ is a prime and $q \ge 1$. Since $3 \cdot a \cdot b = p \cdot q \cdot \gamma$
it must be the case that $p$ divides one of $\{ 3, a, b \}$.


Since $c | (a + b)$ it is also the case that $p | (a + b)$. Suppose $p | a$: Then since $p | (a + b)$, we can conclude that $p | b$ and this
is a contradiction of $(a, b) = 1$. Likewise we reach the same contradiction by supposing $p | b$. So $p$ can only take
on the value $3$. In this case $q | a \cdot b$ which runs into the same contradiction unless $q=1$. This gives $c = 3$ as the
only workable value for $c > 1$. Consequently $c$ is either $1$ or $3.\;\;$ &#x2610;

#### 1.6 Show that if $(a, b)=1$ and $d|(a+b)$ then $(d,a)=(d,b)=1$.

Set $(d,a)=c$. Then $d = c \alpha \textrm{ and } a = c \beta$.

divides both $a$ and $b$. 
This means $(a, b) \ge c$ but we are given that $(a, b) = 1$ so $c = 1 = (d, a)$.

#### 1.10 Given $x$ and $y$, let $m = ax + by$ and $n = cx + dy$ where $ad - bc = \pm 1$. Prove $(m, n) = (x, y)$.


Part 1: Looking ahead we need a general
result, taking $q, r, s > 0$ to avoid absolute value clutter:
If $q|r$ and $q|s$ then $q |(r, s)$. This is easy to 
show: $q|r \implies r = q \alpha$ and likewise $s = q \beta$.
Then $(r, s) = (q \alpha, q \beta) = q (\alpha, \beta)$ from theorem 1.3c; 
so $q|(r,s)$.


Part 2: 
The condition imposed on 
$a, b, c, \textrm{and} \; d$ 
suggests the determinant of a 2D linear transform. 
From linear algebra: The inverse of a $2 \times 2$ matrix
$
A=\left[{\begin{array}{cc}
   a & b \\
   c & d \\
  \end{array}}\right]
\; \; \; $
is
$\; \; \; A^{-1} = \frac{1}{det A} \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]$ with $det A = ad - bc$.
  
  
Writing $x$ and $y$ as a column vector
$\tilde{v}=\left[{\begin{array}{c}x\\y\end{array}}\right]$
and the result of the transform as
$\tilde{k}=\left[{\begin{array}{c}m\\n\end{array}}\right]$
we have $A \tilde{v} = \tilde{k}$ and $\tilde{v}=A^{-1}\tilde{k}$.

$\left[{\begin{array}{c}x\\y\end{array}}\right] =
\frac{1}{det A} \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]
\left[{\begin{array}{c}m\\n\end{array}}\right] =
\pm 1 \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]
\left[{\begin{array}{c}m\\n\end{array}}\right] 
$

In rude mechanical terms
$x = \pm (md-nb) \textrm{ and } y = \pm (na-mc)$.
However we do not need to write this out as such.
Rather say: 
$x$ and $y$ are linear combinations of $m$ and $n$
so $(m, n)|x$ and $(m, n)|y$. This in turn (from part 1)
means that $(m, n)|(x, y)$. 

In the problem statement we see $m$ and $n$ are linear combinations 
of $x$ and $y$. By the same argument then $(x, y)|(m, n)$.
As $(x, y)|(m, n)$ and $(m, n)|(x,y)$ the two are equal,
$(m, n)=(x,y).\;\;$ &#x2610;
