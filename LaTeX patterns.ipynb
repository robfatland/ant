### Delimiters & Sizing

**Auto-sizing** with `\left` / `\right`:

$\begin{align}
\left\lfloor \frac{x}{d} \right\rfloor
\quad
\left\{ \sum_{d|n} \frac{n}{d} \right\}
\end{align}$

**Manual sizing** â€” `\bigl(` `\Bigl(` `\biggl(` `\Biggl(`:

$\bigl( \Bigl( \biggl( \Biggl( \quad \bigl[ \Bigl[ \biggl[ \Biggl[$

**Symbol sizing** â€” ascending: `\small`, `\normalsize`, `\large`, `\Large`, `\LARGE`, `\huge`, `\Huge`:

$\sum \quad {\large \sum} \quad {\Large \sum} \quad {\LARGE \sum} \quad {\huge \sum} \quad {\Huge \sum}$

Wrap in curly braces to compartmentalize: `{\Large \sum}` keeps the sizing local.

Notes:
- `\footnotesize` is **not** recognized in Jupyter's MathJax.
- Nesting size commands does not compound.
- `\relsize` is not supported in Jupyter.

### Equation Layouts

**Inline** (single `$`): $x = y$ â€” compact, flows with text.

**Centered standalone** (double `$$`):

$$x = y$$

**Left-justified standalone** â€” use `$\begin{align} ... \end{align}$`:

$\begin{align}
x = y \iff y = x
\end{align}$

---

**Multi-line, aligned on `&`** (with `\\` for linefeeds):

$\begin{align}
\sigma(n)
= \; & \sigma(2^{a-1}) \cdot \sigma(p) \\
= \; & (2^a-1) \cdot (p + 1) \\
= \; & 2n \\
\end{align}$

**Left-justified sequence** (prefix each line with `&`):

$\begin{align}
& \varphi(n) = n \prod_{p|n} \left(1 - \frac{1}{p}\right) \\
& \sigma(n) = \prod_{p^a \| n} \frac{p^{a+1} - 1}{p - 1} \\
& \mu(n) \in \{-1, 0, 1\}
\end{align}$

---

**Cases (piecewise)**:

$\begin{align}
\mu(n) = \begin{cases}
1 & \textrm{if } n = 1 \\
(-1)^k & \textrm{if } n = p_1 p_2 \cdots p_k \textrm{ (distinct primes)} \\
0 & \textrm{if } p^2 \mid n \textrm{ for some prime } p
\end{cases}
\end{align}$

---

**Stacked conditions under sums** â€” two methods:

Method 1: `\substack` (preferred, cleaner):

$\begin{align}
\sum_{\substack{d \mid n \\ d > 0}} \mu(d) = [n = 1]
\end{align}$

Method 2: `\atop` (older, font shrinks with nesting):

$\begin{align}
\sum_{k=1 \atop (k, d) = 1}^{d} f\Biggl( \frac{k}{d} \Biggr)
\end{align}$

---

**Equation numbering** (manual hack â€” no native support in Jupyter):

$\begin{align}
v + f = e + 2
\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
(1)
\end{align}$

(Pad with `\;` spacers. Not elegant, but it works.)

### Matrices & Vectors

**2Ă—2 matrix**:

$\begin{align}
A=\left[{\begin{array}{cc}
   a & b \\
   c & d \\
  \end{array}}\right]
\quad \textrm{has inverse} \quad
A^{-1} = \frac{1}{\det A} \cdot \left[{\begin{array}{cc}
   d & -b \\
   -c & a \\
  \end{array} }\right]
\end{align}$

**Column vector**:

$\tilde{v}=\left[{\begin{array}{c}x\\y\end{array}}\right]$

**Diagonal matrix** (useful for Dirichlet series manipulations):

$\begin{align}
D = \left[{\begin{array}{ccc}
   \lambda_1 & 0 & 0 \\
   0 & \lambda_2 & 0 \\
   0 & 0 & \lambda_3 \\
  \end{array}}\right]
\end{align}$

### Integrals & Limits

**Definite integral**:

$\begin{align}
\int_{a}^{b} f(x) \; dx = F(b) - F(a)
\end{align}$

**Contour integral** (relevant to analytic number theory):

$\begin{align}
\frac{1}{2\pi i} \oint_{|s|=R} \frac{\zeta(s)}{s} \; ds
\end{align}$

**Limit**:

$\begin{align}
\lim_{x \to \infty} \frac{\pi(x)}{x / \ln x} = 1
\end{align}$

### Raising & Lowering

Push a subscript down with `\lower1ex`:

$\begin{align}
\sigma_{\lower1ex{1 - \alpha}} \quad \textrm{vs} \quad \sigma_{1 - \alpha}
\end{align}$

General positioning: `\raise` and `\lower` with `Nex`:

$\begin{align}
\sigma \; \lower1ex{\sigma} \; \lower2ex{\sigma} \; \lower4ex{\sigma} \; \sigma \; \raise3ex{\sigma} \; \sigma
\end{align}$

### Number Theory Patterns

Common constructions that appear repeatedly in Apostol:

**Dirichlet convolution**:

$\begin{align}
(f * g)(n) = \sum_{d \mid n} f(d) \, g\!\left(\frac{n}{d}\right)
\end{align}$

**MĂ¶bius inversion**:

$\begin{align}
& g(n) = \sum_{d \mid n} f(d) \quad \implies \quad f(n) = \sum_{d \mid n} \mu(d) \, g\!\left(\frac{n}{d}\right)
\end{align}$

**Euler product**:

$\begin{align}
\sum_{n=1}^{\infty} \frac{f(n)}{n^s} = \prod_{p} \left( \sum_{k=0}^{\infty} \frac{f(p^k)}{p^{ks}} \right)
\end{align}$

**Prime factorization with exact powers** (`\|` for "exactly divides"):

$\begin{align}
n = \prod_{i=1}^{k} p_i^{\alpha_i}, \quad p_i^{\alpha_i} \| n
\end{align}$

**Decomposition** $n = a^2 b$ where $b$ is squarefree:

$\begin{align}
a = \prod p_i^{\lfloor \alpha_i / 2 \rfloor}, \quad
b = \prod p_i^{\alpha_i \bmod 2}
\end{align}$

**Summation over coprime residues**:

$\begin{align}
\sum_{\substack{k=1 \\ (k,n)=1}}^{n} f(k)
\end{align}$

**Big-oh and asymptotic notation**:

$\begin{align}
\pi(x) = \frac{x}{\ln x} + \mathcal{O}\!\left(\frac{x}{\ln^2 x}\right)
\end{align}$
