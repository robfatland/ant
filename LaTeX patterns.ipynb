# LaTeX patterns


[Cut to the chase: Reference site link (Oeis)](https://oeis.org/wiki/List_of_LaTeX_mathematical_symbols).


Need to write Möbius or Erdös perhaps?


These notes do inline LaTeX by default. Standalone will give different -- usually nicer -- rendering. The
(simple) syntax for centered standalone is a double-dollar delimiter \$\$ on a line by itself. 
Syntax for left-justified standalone: See below.


Examples:


Inline $x = y$.


Centered:


$$x = y$$


Left-aligned:


$\begin{align}x = y \iff y = x \end{align}$

### Laundry list

- **make it larger**: \\exists < \\Large{\\exists} < \\LARGE{\\exists} : $\exists \;\; \Large{\exists} \;\; \LARGE{\exists}$
    - `\\relsize` is no dice in Jupyter at this time but here is bigness and largeness:
        - \\bigl( \\Bigl( \\Biggl( $\; \rightarrow \; \bigl( \Bigl( \Biggl($
        - \\large( \\Large( \\LARGE( $\; \rightarrow \; \large( \Large( \LARGE($
- **text inclusion** \\textrm{ can be found in }: $x \textrm{ can be found in } y$
- **such that**: \\ni $\ni$ or not as nice \\backepsilon $\backepsilon$
- **element of**: \\in: $a \in B$
- **there exists**: \\exists $\exists$
- **for all**: \\forall $\forall x \in A$
- **logical negation**: \\neg $\neg a$
- **if and only if**: $\iff$
- **AND**, **OR**, **empty set**: \\cap and \\cup $\cap \;\; \cup$ and check out size modifiers: $A \ \bigcap \ B  = \large{\emptyset}$
- add **space**: \\; as in $x y$ becomes $x \; y$. Can also use backslash alone, followed by a space. $x \ y$
- **product** and **sum** notation: \\prod, \\sum: $\prod_{a}^{b} \;\;\; \sum_{c}^{d}$
- Greek **variants**: \\phi versus \\varphi, $\phi \textrm{ versus } \varphi$
- **Set archetypes**: Real numbers etcetera: \\mathbb{R}, \\Large{same} but \\R is a no-go: $\mathbb{R} \;\; \Large{\mathbb{R}}$
- **absolute value, divides etc**: Compare pipe character with \\mid and \\big|: $\;\; | a | \;\;\; \mid a \mid \;\;\; \big| a \big|$
- **Big oh notation using \\mathcal**: $\mathcal{O}(f(x))$

### Integrals


\\int_{a}^{b} f(x) \\; dx = 7


$\begin{align}
\int_{a}^{b} f(x) \; dx = 7
\end{align}$

### Left-justified constructs


Equation sequence, aligned on the \& character with \\\\ end of line indicators: 


$\begin{align}
\sigma(n) 
= \; & \sigma(2^{a-1}) \cdot \sigma(p) \\
= \; & (2^a-1) \cdot (p + 1) \\
= \; & (2^a-1) \cdot (2^a - 1 + 1) \\
= \; & 2 \cdot 2^{a-1} \cdot (2^a - 1) \\
= \; & 2n \\
\end{align}$



<br><br><br>

Multiple conditions


$\begin{align}
\mu(n) = \begin{cases} 
1 & \textrm{ if } n = 1 \\
-1^k & \textrm{ if } n \textrm{ is the product of } k \textrm{ distinct primes } \\
0 & \textrm{ otherwise }
\end{cases}
\end{align}$



<br><br><br>


Using \\atop to center-stack clauses


$\begin{align}
k=1 \atop 
(k, d) = 1 \textrm{ and or perhaps other conditions }
\end{align}$


This mechanism is not implicitly stackable to $n$ clauses; but for what it is worth it
can be nested (with diminishing font size):


$\begin{align}
k=1 \atop 
{
(k, d) = 1 \atop
\textrm{ and or perhaps other conditions such as } d = 19}
\end{align}$


...and of course we can get carried away as in the tale of Fury...


$\begin{align}
k=1 \atop
{(k, d) = 1 \atop
{\textrm{ here we find some additional conditions }  \atop 
\textrm{ like maybe } d = 19
} }
\end{align}$

...so now magnify with \\Large:

$\begin{align}
\Large{
k=1 \atop
{(k, d) = 1 \atop
{\textrm{ here we find some additional conditions }  \atop 
\textrm{ like maybe } d = 19
} } }
\end{align}$



Finally here is the main use: \\atop in a sum (note '\\Biggl('):

            
$\begin{align}
(u * F^{*})(n) = 
\sum_{d|n} F^*(d) = 
\sum_{d|n} \sum_{k=1 \atop (k, d) = 1}^{d} f\Biggl( \frac{k}{d} \Biggr)
\end{align}$
